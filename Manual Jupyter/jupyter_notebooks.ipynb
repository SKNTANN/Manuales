El *Proyecto Jupyter* ([<span style="text-decoration:underline;">Jupyter Proyect</span>](https://jupyter.org "clic aquí")) surge en 2014 como una evolución del proyecto [<span style="text-decoration:underline;">IPython (Interactive Python)</span>](https://ipython.org/ "clic aquí"), una potente consola vitaminada para Python. Sin embargo, Jupyter es mucho más ambicioso que IPython pues pretende construir una plataforma agnóstica del lenguaje que ofrezca a sus usuarios un conjunto de potentes herramientas para trabajar con datos, visualizarlos y poder compartir los resultados.
    
El proyecto fue iniciado por [<span style="text-decoration:underline;">Fernando Pérez</span>](https://bids.berkeley.edu/people/fernando-perez "clic aquí"), un científico y profesor de la Universidad de California Berkeley. El proyecto es software libre y está basado en componentes libres, por lo que está apoyado, desarrollado y mantenido por una amplia comunidad en la que participan tanto profesionales como entusiastas del mismo.

La adopción y el éxito de Jupyter ha sido muy grande, de tal modo que se está convirtiendo en un estandar para el trabajo de los científicos de datos. Hoy en día la herramienta está integrada varias de las plataformas y entornos de desarrollo más importantes tales como [<span style="text-decoration:underline;">Google Cloud</span>](https://cloud.google.com/?hl=es "clic aquí"), [<span style="text-decoration:underline;">Google Colaboratory</span>](https://colab.research.google.com/notebooks/welcome.ipynb "clic aquí"), [<span style="text-decoration:underline;">Microsoft Azure</span>](https://azure.microsoft.com/es-mx/ "clic aquí"), [<span style="text-decoration:underline;">AWS (Amazon Web Services)</span>](https://aws.amazon.com/ "clic aquí"), [<span style="text-decoration:underline;">IMB Bluemix</span>](https://www.ibm.com/cloud-computing/bluemix/es/node/245 "clic aquí"), [<span style="text-decoration:underline;">Databricks Platform</span>](https://databricks.com/ "clic aquí"), [<span style="text-decoration:underline;">Rachspace</span>](https://www.rackspace.com/ "clic aquí"), [<span style="text-decoration:underline;">Continuum</span>](https://www.continuum.net/ "clic aquí"), [<span style="text-decoration:underline;">Jetbrains</span>](https://www.jetbrains.com/ "clic aquí"), entre otros.

## Descarga e Instalación

Hay varias formas de instalar y usar Jupyter, que van desde la instalación a través de [<span style="text-decoration:underline;"><tt style="font-weight:bold;">conda</tt></span>](https://conda.io/ "clic aquí"), [<span style="text-decoration:underline;"><tt style="font-weight:bold;">mamba</tt></span>](https://mamba.readthedocs.io/ "clic aquí"), [<span style="text-decoration:underline;"><tt style="font-weight:bold;">pip</tt></span>](https://pypi.org/project/pip/ "clic aquí"), [<span style="text-decoration:underline;"><tt style="font-weight:bold;">pipenv</tt></span>](https://pipenv.pypa.io/en/latest/ "clic aquí") o [<span style="text-decoration:underline;"><tt style="font-weight:bold;">docker</tt></span>](https://www.docker.com/ "clic aquí").

Para los nuevos usuarios, se recomienda instalar **Anaconda**. Esta plataforma instala convenientemente Python, Jupyter Notebook/Jupyter Lab y otros paquetes de uso común para la informática científica y la ciencia de datos.

Si bien Jupyter ejecuta código en muchos lenguajes de programación, Python es un requisito indispensable (Python >= 3.3) para instalar Jupyter.

<span style="font-size:20px;">Distribución Anaconda</span></center>
<br>

Desde 2011 Python ha incluido el gestor de paquetes [<span style="text-decoration:underline;">pip</span>](https://pypi.org/project/pip/ "clic aquí"), un sistema de administración de paquetes que se usa para instalar y administrar paquetes de software escritos en Python. Sin embargo, para los cálculos numéricos, hay varias dependencias que no están escritas en Python, por lo que las versiones iniciales de pip no pudieron resolver el problema por sí mismas.

Para evitar este problema, [<span style="text-decoration:underline">Continuum Analytics</span>](https://www.anaconda.com/blog/continuum-analytics-officially-becomes-anaconda "clic aquí") lanzó [<span style="text-decoration:underline">Anaconda</span>](https://www.anaconda.com/ "clic aquí"), una distribución libre y abierta para Python, especializada en cómputo científico, ciencia de datos y [<span style="text-decoration:underline">aprendizaje automático</span>](https://es.wikipedia.org/wiki/Aprendizaje_autom%C3%A1tico "clic aquí") que incluye varias plataformas de desarrollo incluso para otros lenguajes de programación como lo son [<span style='text-decoration:underline'>Julia</span>](https://julialang.org/ "clic aquí") y [<span style='text-decoration:underline'>R</span>](https://cran.r-project.org/index.html "clic aquí"). 

También desarrolló [<span style="text-decoration:underline">Conda</span>](https://conda.io/ "clic aquí"), un sistema de administración de paquetes y entornos, que es utilizado por la distribución de Anaconda. Aunque Conda está estrechamente acoplado a Anaconda, los dos son proyectos distintos con diferentes objetivos.

### Descarga

Para instalar Anaconda necesitamos descargar el paquete a través de su página oficial. Anaconda está disponible para Python 3 con instaladores disponibles para Windows, MacOS y Linux en el siguiente enlace: 

1. Ingresar al sitio [<span style="text-decoration:underline;">Anaconda</span>](https://www.anaconda.com/ "clic aquí"). 

2. Hacer clic en la menú "__Products__" -> "__Individual Edition__".
    
    1. *Esta edición permite configurar rápidamente un entorno local simplemente con el paquete de instalación*.
 
 
3. Hacer clic en "__Download__" para iniciar la descarga del instalador. 
    
    1. *Por defecto, el sitio detecta el Sistema Operativo, pero se puede descargar el instalador para otro sistema operativo (o versión) haciendo clic en la opción "__Get Additional Installers__"*.
    
    2. *Para usuarios Windows/MacOS se recomienda descargar el paquete "__Graphical Installer__" para una instalación con interface gráfica*.

### Instalación

En el caso de Windows/MacOS el instalador es similar por lo que los siguientes pasos son similares en ambos sistemas:

1. Ejecutar el instalador de Anaconda descargado. 

2. Hacer clic en "__Next__".

3. Leer los términos de la licencia y clic en "__I Agree__".

4. Seleccionar la instalación "__Just Me__" si la instalación es para un solo usuario.

    - *En Windows, si el software será usado por otros usuarios, seleccionar "__All Users__" (requiere privilegios de administrador)*.
    
5. Seleccionar el directorio por defecto (recomendado) de destino para instalar Anaconda.

6. En Windows, elegir la opción por defecto "__Register Anaconda3 as my default Python 3.x__" (recomendado). A menos que se planee instalar y ejecutar múltiples versiones de Anaconda o múltiples versiones de Python, aceptar el valor predeterminado y dejar esta casilla marcada.
 
    - *No se recomienda elegir la opción "__Add Anaconda3 to my PATH enviroment variable__". Esta opción puede interferir con otro software*. 
    
7. Clic en "__Install__" para instalar Anaconda. 

    - *Opcional: Se puede instalar opcionalmente [<span style="text-decoration:underline;">PyCharm</span>](https://www.jetbrains.com/es-es/pycharm/ "clic aquí") para Anaconda en el enlace proveído*.

## Jupyter Server

*Jupyter Server* proporciona el *backend* (es decir, los servicios principales, las [<span style="text-decoration:underline;">API</span>](https://en.wikipedia.org/wiki/API "clic aquí") y los puntos finales [<span style="text-decoration:underline;">REST</span>](https://en.wikipedia.org/wiki/Representational_state_transfer "clic aquí")) para aplicaciones web.

Una aplicación web es una herramienta basada en navegador web para la creación interactiva de documentos que combinan texto, matemáticas, cálculos y multimedia enriquecida.

El Proyecto Jupyter actualmente implementa dos aplicaciones web para la computación interactiva: 

- [<span style="font-weight:bold;text-decoration:underline;">Jupyter Notebook</span>](https://jupyter-notebook.readthedocs.io/en/stable/ "clic aquí")

- [<span style="font-weight:bold;text-decoration:underline;">JupyterLab</span>](https://jupyterlab.readthedocs.io/en/stable/ "clic aquí")

Las aplicaciones web de Jupyter se basan en una [<span style="text-decoration:underline;">estructura servidor-cliente</span>](https://en.wikipedia.org/wiki/Client%E2%80%93server_model "clic aquí"). El servidor del notebook utiliza una arquitectura de [<span style="text-decoration:underline;font-style:italic;">kernel de dos procesos</span>](https://ipython.readthedocs.io/en/stable/overview.html#ipythonzmq "clic aquí") basada en [<span style="text-decoration:underline;">ZeroMQ</span>](https://zeromq.org/ "clic aquí"), así como [<span style="text-decoration:underline;">Tornado</span>](https://www.tornadoweb.org/en/stable/ "clic aquí") para atender [<span style="text-decoration:underline;">solicitudes HTTP</span>](https://www.ibm.com/docs/en/cics-ts/5.3?topic=protocol-http-requests "clic aquí"):

Al ejecutar un servidor de Jupyter este se crea en la máquina local que lo ejecuta y no requiere necesariamente de una conexión a Internet. Luego, el servidor de Jupyter abre el *cliente* (Jupyter Notebook o JupyterLab) en el navegador web predeterminado.

De forma predeterminada, un servidor de Jupyter se ejecuta localmente en la [<span style="text-decoration:underline;">dirección IP</span>](https://en.wikipedia.org/wiki/IP_address "clic aquí") "__127.0.0.1:8888__" y solo es accesible desde [<span style="text-decoration:underline;">localhost</span>](https://en.wikipedia.org/wiki/Localhost "clic aquí"). Para acceder al servidor de Jupyter, se puede abrir desde cualquier navegador utilizando: __http://127.0.0.1:8888__ o __http://localhost:8888__.

La mayoría de las veces no se necesitará iniciar Jupyter Server directamente. Las aplicaciones web de Jupyter (como Jupyter Notebook, Jupyterlab, [<span style="text-decoration:underline;">Voilà</span>](https://voila.readthedocs.io/en/stable/index.html "clic aquí"), etc.) vienen con sus propios puntos de entrada que inician un servidor automáticamente.

### Kernels 

Los *Kernels* (núcleos) son procesos específicos del lenguaje de programación que se ejecutan de forma independiente e interactúan con las aplicaciones de Jupyter y sus interfaces de usuario. [<span style="text-decoration:underline;">IPython</span>](https://ipython.org/ "clic aquí") es el kernel de Jupyter por defecto, que proporciona un entorno potente para la informática interactiva en Python.

En términos generales, un "kernel" es un programa que se ejecuta e introspecta el código del usuario. IPython incluye un kernel para código Python, pero otros desarrolladores han escrito kernels para varios otros lenguajes.

Cada kernel es un motor de ejecución para un lenguaje o plataforma concreta que ejecuta en el servidor. A través de un [<span style="text-decoration:underline;">sistema de colas</span>](https://www.um.es/or/ampliacion/node3.html "clic aquí") basado en [<span style="text-decoration:underline;">ZMQ</span>](https://zeromq.org/ "clic aquí") el código ejecutado en un notebook es enviado al kernel correspondiente.

Actualmente Jupyter soporta integración con más de [<span style="text-decoration:underline;">40 lenguajes de programación</span>](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels "clic aquí") de los que destacan: [<span style="text-decoration:underline;">Python</span>](https://www.python.org/), [<span style="text-decoration:underline;">Julia</span>](https://julialang.org/), [<span style="text-decoration:underline;">R</span>](https://cran.r-project.org/index.html) , [<span style="text-decoration:underline;">Scala</span>](https://www.scala-lang.org/), [<span style="text-decoration:underline;">Ruby</span>](https://www.ruby-lang.org/es/) o [<span style="text-decoration:underline;">Go</span>](https://golang.org/). Pero también es fácilmente integrable con herramientas y plataformas de [<span style="text-decoration:underline;">Big Data</span>](https://en.wikipedia.org/wiki/Big_data "clic aquí") como [<span style="text-decoration:underline;">Spark</span>](https://spark.apache.org), lo que permite abstraerse de la complejidad de estas herramientas, aprovechando todo su potencial desde un entorno muy amigable. 

Cada Kernel tiene una forma particular de instalación. Para obtener más detalles acerca de la instalación de otros kernels se puede consultar la documentación oficial de Jupyter (*en inglés*) en el siguiente enlace: [<span style="text-decoration:underline;">Kernel install</span>](https://ipython.readthedocs.io/en/latest/install/kernel_install.html "clic aquí")

# Creando un notebook

En esta sección se revisará cómo crear un notebook, cómo usar las celdas de código y cómo formatear texto en celdas de marcado usando [<span style="text-decoration:underline;">lenguaje de marcado</span>](https://es.wikipedia.org/wiki/Lenguaje_de_marcado "clic aquí"). 

A partir de esta sección se usará a JupyterLab para crear y trabajar notebooks así como para otras actividades relacionadas a este propósito. 

Para crear un notebook en JupyterLab se puede hacer de dos formas:

1. Desde el menú "__File__" -> "__New__" -> "__Notebook__".

2. Desde la pestaña "__Launcher__" -> "__Notebook (Python 3)__". 

Cualquiera de las dos formas anteriores abrirá un nuevo notebook con el nombre *untitled.ipynb*, el cual será creado en el directorio actual de trabajo. Para cambiar el nombre del notebook se puede hacer desde el menú "__File__" -> "__Rename Notebook...__". 

## Celdas de código

Las celdas de código en un notebook están destinada exclusivamente a ejecutar código de algún lenguaje de programación asociado a un kernel. 

Por defecto, el kernel de excelencia que se puede iniciar junto con un notebook es el kernel "*ipykernel (Python 3)*", lo que significa que las celdas de código podrán ejecutar código basado en Python 3. 

En las celdas de código se pueden ejecutar código de una sola línea, fragmentos de código (varias líneas) o incluso programas completos. Además, se pueden ejecutar comandos para ejecutar tareas del sistema. 

### Ejecutando código

Para ejecutar código o comandos en las celdas de código se puede hacer de tres distintas formas:

1. Menú "__Run__" -> "__Run Selected Cells__"

2. Desde la barra de herramientas haciendo clic en el ícono: <kbd style="font-weight:bold;">▶</kbd>

3. Mediante la combinación de teclas <kbd style="font-weight:bold;">Shift &#8679;</kbd> + <kbd style="font-weight:bold;">Enter &#9166;</kbd>

Por ejemplo, el siguiente fragmento de código es una *definición de función* en Python 3 y la cual realiza una serie de operaciones aritméticas dados dos números:

```python
def operaciones(num_1, num_2):
    suma = num_1 + num_2
    resta = num_1 - num_2
    mul = num_1 * num_2
    div = num_1 / num_2
    print(f"Suma de {num_1} y {num_2} es {suma}",\
          f"Resta de {num_1} y {num_2} es {resta}",\
          f"Multiplicación de {num_1} y {num_2} es {mul}",\
          f"División de {num_1} y {num_2} es {div}", sep='\n')
```

Entonces, para poder ejecutar y usar esta función basta con escribirla (o copiarla directamente del ejemplo anterior) y luego ejecutar la celda con alguna de las opciones anteriomente mencionadas:

````python
def operaciones(num_1, num_2):
    suma = num_1 + num_2
    resta = num_1 - num_2
    mul = num_1 * num_2
    div = num_1 / num_2
    return (f"Suma de {num_1} y {num_2} es {suma}",\
            f"Resta de {num_1} y {num_2} es {resta}",\
            f"Multiplicación de {num_1} y {num_2} es {mul}",\
            f"División de {num_1} y {num_2} es {div}")
````

````python
operaciones(3,5)
````

También hay ocasiones que por diversos factores los programas entran en un [<span style="text-decoration:underline;">bucle</span>](https://es.wikipedia.org/wiki/Bucle_(programaci%C3%B3n) "clic aquí") o ciclo "infinito" lo que hace imposible que otras celdas subsecuentes se ejecuten.

Para reiniciar el kernel activo en un notebook se puede hacer de las siguientes maneras:

1. Menú "__Kernel__" -> "__Restart Kernel...__"

2. Desde la barra de herramientas haciendo clic en el ícono: <kbd style="font-weight:bold;font-size:15px">⟳</kbd>

### Lenguajes de marcado

Un **lenguaje de marcado o lenguaje de marcas** es una forma de codificar un documento que, junto con el texto, incorpora "etiquetas" o "marcas" que contienen información adicional acerca de la estructura del texto o su presentación original del archivo.

En otras palabras, permite convertir texto plano en texto formateado visualmente más presentable. Actualmente el lenguaje de marcado más extendido en el mundo es HTML. 

Por otro lado, un **lenguaje de marcado ligero** es un tipo de formateo de texto más o menos estandarizado, que ocupa poco espacio y es fácil de editar. En algunas ocasiones los lenguajes de marcado [<span style="text-decoration:underline;">XML</span>](https://en.wikipedia.org/wiki/XML "clic aquí") o [<span style="text-decoration:underline;">HTML</span>](https://en.wikipedia.org/wiki/HTML "clic aquí"), hacen demasiado complejo y grande un archivo, sin aportar ninguna o casi ninguna ventaja respecto a un lenguaje más simple. Ejemplos de lenguajes de marcado lijero son [<span style="text-decoration:underline;">Markdown</span>](https://en.wikipedia.org/wiki/Markdown "clic aquí") o [<span style="text-decoration:underline;">ReStructuredText</span>](https://en.wikipedia.org/wiki/ReStructuredText "clic aquí"). 

Aunque Markdown es el lenguaje de marcas predilecto para emplear en celdas de marcado, se utiliza HTML para cualquier marca que no esté cubierta por la sintaxis de Markdown.

#### Markdown

**Markdown** es un conjunto simple de reglas para diseñar y formatear texto sin formato. Esta diseñado para permitir escribir utilizando un formato de texto plano fácil de leer y escribir, y luego convertirlo en XHTML (o HTML) estructuralmente válido. 

Por tanto, Markdown es dos cosas: una sintaxis de formato de texto plano; y una herramienta de software, escrita en [<span style="text-decoration:underline;">Perl</span>](https://es.wikipedia.org/wiki/Perl "clic aquí"), que convierte el formato de texto plano en HTML. 

El objetivo primordial del diseño de la sintaxis de formato de Markdown es hacerla lo más legible posible. La idea es que un documento con formato Markdown sea publicable tal cual, como texto plano, sin que parezca que ha sido marcado con etiquetas o instrucciones de formato. Aunque la sintaxis de Markdown se ha visto influenciada por varios filtros existentes de conversión de texto a HTML, la mayor fuente de inspiración para la sintaxis de Markdown es el formato del correo electrónico en texto plano.

#### HTML

**HTML** (HyperText Markup Language) es un lenguaje de marcado y es el elemento más básico de la Web. Define el significado y la estructura del contenido web. Además de HTML, generalmente se utilizan otras tecnologías para describir la apariencia/presentación de una página web ([<span style="text-decoration:underline;">CSS</span>](https://en.wikipedia.org/wiki/CSS "clic aquí")) o la funcionalidad/comportamiento ([<span style="text-decoration:underline;">JavaScript</span>](https://en.wikipedia.org/wiki/JavaScript "clic aquí")).

El lenguaje HTML utiliza "marcas" para anotar texto, imágenes y otros contenidos para su visualización en un navegador web. El marcado HTML se aplica mediante *etiquetas* de apertura y cierre. El nombre de una etiqueta determina la marca que se aplicará a un determinado elemento. Por ejemplo, para marcar un texto en "negrita" se usan las etiquetas "<tt style="font-weight:bold;">&lt;b&gt;...&lt;/b&gt;</tt>":

```html
<b>Texto en negrita</b>
```

Aquí, el nombre de la etiqueta (__b__) hace referencia a la palabra inglesa *bold*. El nombre de las etiquetas no distingue entre mayúsculas y minúsculas. Así, por ejemplo, la etiqueta <tt style="font-weight:bold;">&lt;title&gt;</tt> puede escribirse como <tt style="font-weight:bold;">&lt;Title&gt;</tt> o <tt style="font-weight:bold;">&lt;TITLE&gt;</tt>, también.

Además, algunas etiquetas admiten *atributos* los cuales proveen información adicional acerca de los elementos. Los atributos siempre son expecificados en la etiqueta de apertura y usualmente vienen en pares *nombre="valor"*. Por ejemplo, para crear un [<span style="text-decoration:underline;">hipervínculo</span>](https://en.wikipedia.org/wiki/Hyperlink "clic aquí") se utilizan las etiquetas "<tt style="font-weight:bold;">&lt;a&gt;...&lt;/a&gt;</tt>" y el atributo "<tt style="font-weight:bold;">href</tt>" para especificar la [<span style="text-decoration:underline;">URL</span>](https://en.wikipedia.org/wiki/URL "clic aquí") del recurso enlazado:

```html
<a href="https://www.google.com/">Enlace a Google</a> 
```

#### Markdown vs HTML

A pesar de que Markdown fue desarrollado para permitir que las personas escriban con un formato de texto plano fácil de leer y de escribir, no es un sustituto de HTML ni se acerca a él. Su sintaxis es mínima y solo se correlaciona con una pequeña proporción de etiquetas HTML. La idea de Markdown es facilitar la lectura, escritura y edición de la prosa sin la intención de crear una sintaxis que solo sea para agregar rápidamente etiquetas HTML. 

Así, la sintaxis de formato de Markdown aborda solo los problemas que se pueden expresar en texto sin formato. Mientras que se recurre a HTML para cualquier marcado que no esté cubierto por la sintaxis de Markdown. 

En conclusión, **HTML es un formato para publicación mientras que Markdown es un formato para lectura**.

# Formatos de texto

*¿Por qué no hay menús desplegables de selección de fuente y tamaño de fuente, botones que pueda presionar para poner en negrita y cursiva mi texto u otras opciones de estilo avanzadas en un notebook?*

Cuando se usa [<span style="text-decoration:underline;">Microsoft Word</span>](https://es.wikipedia.org/wiki/Microsoft_Word "clic aquí"), [<span style="text-decoration:underline;">Google Docs</span>](https://en.wikipedia.org/wiki/Google_Docs "clic aquí"), [<span style="text-decoration:underline;">Apple Pages</span>](https://es.wikipedia.org/wiki/Pages "clic aquí"), [<span style="text-decoration:underline;">Apache OpenOffice Writer</span>](https://es.wikipedia.org/wiki/Apache_OpenOffice_Writer "clic aquí") o cualquier otro software de procesamiento de texto, generalmente se usa el mouse para seleccionar varias opciones de estilo, como interlineado, tamaño de fuente, color de fuente, formato de párrafo, etc. El sistema a menudo se describe como una interfaz "*lo que ves es lo que obtienes*". Esto significa que la entrada (lo que se le dice a la computadora) coincide exactamente con la salida (lo que la computadora devuelve). 

Sin embargo, un notebook no utiliza una interfaz que interactúe directamente con "botones" para aplicar formato a texto. En su lugar, utiliza algo llamado "lenguaje de marcado".

Markdown es, precisamente, un lenguaje de marcado que genera HTML y que la celda (de tipo *Markdown*) puede interpretar y representar. Esto significa que las celdas de Markdown también pueden representar código HTML sin formato. 

Entonces, *¿Por qué utilizar Markdown o HTML?*. Es muy importante que las personas que aprenden y usan lenguajes de programación aprendan a conceptualizar la **entrada** y la **salida** como dependientes, pero también distintas. Una buena razón para utilizar Markdown es que fomenta este tipo de pensamiento. Es de reconocerse que Markdown, a diferencia de HTML, es algo limitado en sus opciones de estilo, lo que significa que hay inherentemente menos preocupaciones no específicas del tema por las que agonizar mientras se trabaja. El concepto de esta filosofía es que, al usar Markdown, se piensa en el contenido de la escritura y no necesariamente en su formato y apariencia.

## Formato de texto básico

A diferencia de otros procesadores de texto como Google Docs, Microsoft Word o Write de Open Office en los cuales podemos formatear texto a través de "accesos directos" en las celdas Markdown de los notebook de Jupyter para poder formatear o aplicar *estilo* al texto plano que escribimos en ellas se usan los [lenguajes de marcado](#Lenguajes-de-marcado). 

### Énfasis

Para denotar énfasis en un texto generalmente se emplea estilos como "negritas", "italicas", "subrayado", "tachado" o alguna combinación de ellas. 

En Markdown, para que un texto se muestre en letra itálica (cursiva) se emplea un asterísco (o un guión bajo) tanto al principio como al final del texto a marcar:

```markdown
*texto en cursiva*
_texto en cursiva_
```
> *texto en cursiva*

Usando doble asteríscos o dos guiones bajos se logra un formato en negritas:

```markdown
**texto en negritas**
__texto en negritas__
```
> **texto en negritas**

Usando tres asteriscos o guiones bajos genera un formato combinado de negritas y cursivas:

```markdown
***texto en negritas y cursiva***
___texto en negritas y cursiva___
```
> ***texto en negritas y cursiva***

Para aplicar tachado a un texto requiere emplear dos virgulillas (<tt style="font-weight:bold;">~</tt>):

```markdown
~~texto tachado~~
```
> ~~texto tachado~~

Este último formato también se puede combinar con los anteriores apicando la sintaxis adecuada donde sea requerida, por ejemplo:

```markdown
***~~texto en negritas, cursiva y tachado~~***
```
> ***~~texto en negritas, cursiva y tachado~~***

Por otra parte, para lograr el mismo tipo de formato empleando HTML se emplean las etiquetas "<tt style="font-weight:bold;">i</tt>", "<tt style="font-weight:bold;">b</tt>" y "<tt style="font-weight:bold;">s</tt>" para cursivas, negritas y tachado, respectivamente:

```html
<i>texto en cursiva</i>
<b>texto en negritas</b>
<s>texto tachado</s>
```
> <i>texto en cursiva</i><br>
<b>texto en negritas</b><br>
<s>texto tachado</s>

De igual forma que en Markdown también en HTML se pueden anidar las etiquetas para crear una combinación de formatos al mismo texto:

```html
<i><b><s>texto en cursiva, negritas y tachado</s></b></i>
```
> <i><b><s>texto en cursiva, negritas y tachado</s></b></i>

Además de los estilos de énfasis vistos anteriormente hay otros formatos de énfasis que a menudo pueden ser útiles para resaltar ciertas partes de un texto. En HTML las etiquetas "<tt style="font-weight:bold;">u</tt>" y "<tt style="font-weight:bold;">mark</tt>" se emplean para subrayar y resaltar texto, respectivamente:

```html
<u>Texto subrayado</u>
<mark>Texto resaltado</mark>
```
> <u>Texto subrayado</u>><br>
> <mark>Texto resaltado</mark>

Desafortunadamente no hay una sintaxis en Markdown equivalente que se pueda emplear para lograr el mismo formato anterior.

### Tamaño

Cambiar el tamaño de fuente es una tarea común al aplicar formato a un texto. En Markdown no existe alguna manera de modificar el tamaño de fuente, pero a través de la sintaxis de HTML sí es posible lograrlo. 

En HTML existe la etiqueta "<tt style="font-weight:bold;">&lt;font&gt;</tt>" que permite definir un tamaño de fuente fijo. Solo se puede definir siete tipos de tamaño de fuente fijo a través del atributo "<tt style="font-weight:bold;">size</tt>" que acepta valores numéricos entre el 1 y 7, donde 1 es el tamaño de fuente más pequeño y 7 el más grande:

```html
<font size=1>Tamaño de fuente 1</font>
<font size=7>Tamaño de fuente 7</font>
```
> <font size=1>Tamaño de fuente 1</font>

> <font size=7>Tamaño de fuente 7</font>

### Color

Otro formato de texto común es cambiar el color de la fuente. En Markdown no hay una sintaxis para definir el color de fuente, pero es posible cambiar el color de fuente usando HTML mediante la etiqueta "<tt style="font-weight:bold;">&lt;font&gt;</tt>" y su atributo "<tt style="font-weight:bold;">color</tt>":

```html
<font color="red">Texto rojo</font>
<font color="blue">Texto azul</font>
<font color="yellow">Texto amarillo</font>
```
> <font color="red">Texto rojo</font><br>
<font color="blue">Texto azul</font><br>
<font color="yellow">Texto amarillo</font>

El atributo "<tt style="font-weight:bold;">color</tt>", además de aceptar valores por nombre, también acepta valores [<span style="text-decoration:underline;">hexadecimales (HEX)</span>](https://es.wikipedia.org/wiki/Sistema_hexadecimal "clic aquí").

Por ejemplo, el equivalente en valores hexadecimales de los casos anteriores sería: 

```html
<font color="#ff0000">Texto rojo</font>
<font color="#0000ff">Texto azul</font>
<font color="#ffff00">Texto amarillo</font>
```

> <font color="#ff0000">Texto rojo</font><br>
<font color="#0000ff">Texto azul</font><br>
<font color="#ffff00">Texto amarillo</font>

Para obtener los valores hexadecimales de distintas combinaciones de colores se puede encontrar una referencia en el siguiente enlace:

- [Colores HEX](https://www.w3schools.com/html/html_colors_hex.asp "clic aquí")

### Alineación 

En HTML es posible alinear párrafos empleando la etiqueta "<tt style="font-weight:bold;">&lt;p&gt;</tt>" (que hace referencia a la palabra inglesa *paragraph*) la cual acepta el atributo "<tt style="font-weight:bold;">style</tt>". En este atributo se define la propiedad "<tt style="font-weight:bold;">text-align</tt>" en la cual se establece el valor correspondiente de acuerdo al tipo de alineación:

- <tt style="font-weight:bold;">left</tt>: alineación izquierda.
- <tt style="font-weight:bold;">right</tt>: alineación derecha.
- <tt style="font-weight:bold;">justify</tt>: alineación justificada.
- <tt style="font-weight:bold;">center</tt>: alineación centrada.

Dicho lo anterior, los siguientes ejemplos clarifican el uso de estas propiedades:

```html
<p style="text-align:left">Texto alineado a la izquierda.</p>
<p style="text-align:right">Texto alineado a la derecha.</p>
<p style="text-align:justify">Texto justificado.</p>
<p style="text-align:center">Texto centrado.</p>
```
> <p style="text-align:left">Texto alineado a la izquierda.</p>
> <p style="text-align:right">Texto alineado a la derecha.</p>
> <p style="text-align:justify">Texto justificado.</p>
> <p style="text-align:center">Texto centrado.</p>

Desafortunadamente no hay una sintaxis equivalente en Markdown para alinear texto.

### Estilos de formato

Como se ha visto hasta ahora el lenguaje de marcado HTML ofrece mayores capacidades para poder definir el formato a un texto. Una de las ventajas de las etiquetas de HTML es que se puede definir varios tipos de formato usando el atributo global "<tt style="font-weight:bold;">style</tt>". 

Por ejemplo, en la etiqueta <tt style="font-weight:bold;">&lt;span&gt;</tt> se puede definir una serie de propiedades a través del atributo "<tt style="font-weight:bold;">style</tt>" con las cuales se pueden establecer distintos formatos a un mismo texto. 

En el siguiente caso, por ejemplo, se cambia el color y tamaño de fuente con la siguiente sintaxis: 

```html
<span style="color:red; font-size:20px">Texto de ejemplo</span>
```

> <span style="color:red; font-size:20px">Texto de ejemplo</span>


En el ejemplo anterior, las propiedades del atributo "<tt style="font-weight:bold;">style</tt>" se definen como un par "*propiedad:valor*" y cada propiedad se separa mediante punto y coma (__;__). En el caso del color de fuente se define la propiedad y valor "<tt style="font-weight:bold;">color:red</tt>" para aplicar color rojo al texto; en el caso del tamaño de fuente se define la propiedad y valor "<tt style="font-weight:bold;">font-size:20px</tt>" para aplicar un tamaño de fuente de 20 *pixeles*. 

Otras propiedades se pueden definir para aplicar subrayado, negritas o incluso color de marcado a un mismo texto usando la misma etiqueta y atributo como en el siguiente ejemplo: 

```html
<span 
style="color:red; 
text-decoration:underline;
font-style: italic;
font-weight: bold;
font-size: 25px;
background-color: #FFFF00">Texto de ejemplo</span>
```
> <span style="color: red; text-decoration: underline; font-style: italic; font-weight: bold; font-size: 25px; background-color: #FFFF00;">Texto de ejemplo</span>

En este ejemplo, en orden se aplican las siguiente propiedades y valores:

1. <tt><b>color:red</b></tt>: color de texto -> rojo

2. <tt><b>text-decoration:underline</b></tt>: decoración de texto -> subrayado

3. <tt><b>font-style:italic</b></tt>: estilo de fuente -> itálica

4. <tt><b>font-weight:bold</b></tt>: marcado de fuente -> negritas

5. <tt><b>font-size:10px</b></tt>: tamaño de fuente -> 25 pixeles 

6. <tt><b>background-color:#FFFF00</b></tt>: color de fondo -> amarillo (<tt style="font-weight:bold;">#FFFF00</tt>)

Cada una de las propiedades puede aceptar otros valores distintos a los del ejemplo. También es posible definir y agregar otras propiedades para modificar el formato de texto. 

Para obtener más detalles acerca de la etiqueta "<tt style="font-weight:bold;">&lt;span&gt;</tt>" se puede consultar el siguiente enlace: [<span style="text-decoration:underline;">etiqueta <tt style="font-weight:bold;">&lt;span&gt;</tt></span>](https://developer.mozilla.org/es/docs/Web/HTML/Element/span "clic aquí")

## Encabezados

Tanto en Markdown como en HTML se pueden crear encabezados (títulos) y en ambos casos sólo existen *seis niveles de profundidad*, donde el primer nivel corresponde al encabezado más grande y el sexto nivel al encabezado más pequeño. 

En Markdown, se usa el símbolo almohadilla o numeral "<tt><b>#</b></tt>" para crear encabezados y el número de almohadillas determina el nivel del encabezado. Por ejemplo, para crear un encabezado de primer nivel se usa una sola almohadilla:

```markdown
# Encabezado nivel 1
```
> <font style='font-size:29px'>Encabezado nivel 1</font>


Y para un encabezado de segundo nivel se usan dos almohadillas:

```markdown
## Encabezado nivel 2
```
> <font style='font-size:24px'>Encabezado nivel 2</font>

Es importante notar que el nombre del encabezado siempre está separado por al menos un espacio de la almohadilla. De no harcerlo, el formato del texto no se aplicará.  

Por otro lado, en HTML los encabezados se denotan a través de las etiquetas "<tt style="font-weight:bold;">h</tt>" que hacen referencia a la palabra inglesa *heading*. Estas etiquetas se acompañan con un número del 1 al 6 que indica el nivel de profundidad del encabezado. En el caso de HTML se tiene que usar una etiqueta de apertura y otra de cierre y, en medio de ellas, el texto al que se aplicará el formato. 

Por ejemplo, crear un encabezado de primer nivel en HTML requeriría de la sintaxis singuiente:

```html
<h1>Encabezado nivel 1</h1>
```
> <font style='font-size:29px'>Encabezado nivel 1</font>

Y para un encabezado de nivel dos:

```html
<h2>Encabezado nivel 2</h2>
```
> <font style='font-size:24px'>Encabezado nivel 2</font>

Hay que tener en cuenta que la etiqueta de cierre siempre lleva una diagonal (<tt style="font-weight:bold;">/</tt>) antes del nombre de la etiqueta.

## Salto de sección y línea

Los saltos de sección y de línea usualmente son empleados en un texto para dar estructura al mismo. Ya sea para crear párrafos o definir una nueva sección se deben emplear ciertos elementos para dar coherencia al texto que se está escribiendo. 

### Salto de sección

En un documento normalmente se usan *saltos de sección* para definir secciones que definen diferentes formatos de forma independiente. Se puede definir visualmente un salto de sección en un notebook usando líneas horizontales. Tanto en Markdown como en HTML es posible crear una línea que vaya a lo largo de toda la celda. 

En el caso de Markdown para crear una línea horizontal se deben colocar tres o más asteriscos, guiones medios o guiones bajos:

```markdown
---

Sección 1
___

Sección 2

***
```
> ---
>
> Sección 1
>___
>
> Sección 2
>
> ***

Con la sintaxis de HTML se pueden agregar líneas horizontales usando la etiqueta "<tt style="font-weight:bold;">&lt;hr&gt;</tt>": 

```html
<hr>
Sección 1
<hr>
Sección 2
<hr>
```
> <hr>
> Sección 1
> <hr>
> Sección 2
> <hr>

Cabe destacar que no es necesario agregar una etiqueta de cierre para crear una línea horizontal. Además, la etiqueta <tt style="font-weight:bold;">&lt;hr&gt;</tt> acepta múltiples atributos para poder personalizar el estilo de línea. Por ejemplo:

```html
<hr width=500px size=10px, align=center>
```
<hr width=500px size=10, align=center>

En este caso, la línea horizontal solo tiene 500px (pixeles) de ancho, 10 px de alto y está alineada al centro. Otra forma más práctica de personalizar el estilo de la línea horizontal es usando el atributo "<tt style="font-weight:bold;">style</tt>" para agregar propiedades adicionales:

```html
<hr style="width:500px; height:5px; color:red; background-color:blue">
```
<hr style="width:500px; height:5px; color:red; background-color:blue">

### Salto de línea

En Markdown se puede generar un *salto de línea* simplemente pulsando la tecla <kbd style="font-weight:bold;">Enter &#9166;</kbd> y dando dos espacios con la barra espaciadora para pasar a la siguiente línea. Para definir un nuevo párrafo solo se tiene que pulsar dos veces la tecla <kbd style="font-weight:bold;">Enter &#9166;</kbd>:

```markdown
Texto Texto Texto Texto Texto Texto Texto  
Texto Texto Texto Texto Texto Texto Texto  
Texto Texto Texto Texto Texto Texto Texto  

Texto Texto Texto Texto Texto Texto Texto  
Texto Texto Texto Texto Texto Texto Texto
```

> Texto Texto Texto Texto Texto Texto Texto  
> Texto Texto Texto Texto Texto Texto Texto  
> Texto Texto Texto Texto Texto Texto Texto  
>
> Texto Texto Texto Texto Texto Texto Texto  
> Texto Texto Texto Texto Texto Texto Texto


Por el contrario, en HTML definir un salto de línea requiere de la etiqueta "<tt style="font-weight:bold;">&lt;br&gt;</tt>".  Consideremos el siguiente ejemplo:

```html
<h1>Titulo</h1>
<br>
<br>
<br>
Texto Texto Texto Texto Texto Texto Texto<br>
Texto Texto Texto Texto Texto Texto Texto<br>
Texto Texto Texto Texto Texto Texto Texto<br>
```
> <span style='font-size:29px'>Titulo</span>
> <br>
> <br>
> <br>
> Texto Texto Texto Texto Texto Texto Texto<br>
> Texto Texto Texto Texto Texto Texto Texto<br>
> Texto Texto Texto Texto Texto Texto Texto<br>

También es posible anidar las etiquetas <tt style="font-weight:bold;">&lt;br&gt;</tt> para ahorrar espacio:

```html
<h1>Titulo</h1>
<br><br><br>
Texto Texto Texto Texto Texto Texto Texto<br>
Texto Texto Texto Texto Texto Texto Texto<br>
Texto Texto Texto Texto Texto Texto Texto<br>
```

> <span style='font-size:29px'>Titulo</span>
> <br><br><br>

## Citas en bloque

En Markdown se pueden generar bloques que contenga texto para crear citas. Es posible usar esta característica con el carácter "__>__" (mayor que) seguido del texto a citar. Al crear una cita en bloque se agrega una barra vertical del lado izquierdo del texto citado del mismo tamaño que la amplitud del texto. 

Hay dos estilos para citar un texto: *estandar* y *anidado*.

### Cita estándar

Una cita estándar simplemente es una porción de texto a citar resaltado por una una única barra vertical *indentada*<sup style='font-weight:bold'><a href='#1'>1</a></sup> (sangría izquierda). Se puede crear una cita usando un solo carácter "__>__" al principop del texto a citar:

```markdown
> Yo llegué en el momento justo, interesado, cuando el hipertexto e
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran.  
Tim Berners-Lee, *Tejiendo la Web*.
```
> Yo llegué en el momento justo, interesado, cuando el hipertexto e
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran.  
Tim Berners-Lee, *Tejiendo la Web*.

Si el texto está compuesto por varias líneas como en el ejemplo anterior, se tiene que dejar un solo salto de línea entre cada línea. Usar más de un salto de línea en una cita dejará fuera de la cita la siguiente línea:

```markdown
> Yo llegué en el momento justo, interesado, cuando el hipertexto e
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran.  

Tim Berners-Lee, *Tejiendo la Web*.
```
> Yo llegué en el momento justo, interesado, cuando el hipertexto e
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran.  

Tim Berners-Lee, *Tejiendo la Web*.

Si es explícitamente necesario agregar un más de un salto de línea en una cita de varias líneas, se puede resolver el problema usando el carácter "__>__" en cada una de las líneas del texto citado, incluso si es una línea en blanco:

También es posible agregar más de un carácter para citar multiples líneas en un solo bloque:

```markdown
>Y o llegué en el momento justo, interesado, cuando el hipertexto e Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran.
>
> Tim Berners-Lee, *Tejiendo la Web*.
```
> Yo llegué en el momento justo, interesado, cuando el hipertexto e Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran.
>
> Tim Berners-Lee, *Tejiendo la Web*.

Por otro lado, para crear citas en HTML se emplea la etiqueta "<tt style="font-weight:bold;">&lt;blockquote&gt;</tt>":

```html
<blockquote>
Yo llegué en el momento justo, interesado, cuando el hipertexto e 
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran. <br><br>
Tim Berners-Lee, *Tejiendo la Web*.
</blockquote>
```
<blockquote>
Yo llegué en el momento justo, interesado, cuando el hipertexto e 
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran. <br><br>
Tim Berners-Lee, *Tejiendo la Web*.
</blockquote>

También es posible usar la forma equivalente usando la etiqueta "<tt style="font-weight:bold;">&lt;q&gt;</tt>". Esta etiqueta a diferencia de "<tt style="font-weight:bold;">&lt;blockquote&gt;</tt>" no emplea una barra vertical para denotar una cita sino que en su lugar emplea dobles comillas para resaltar la cita y sin aplicar sangria a la cita:

```html
<q>
Yo llegué en el momento justo, interesado, cuando el hipertexto e 
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran. <br><br>
Tim Berners-Lee, *Tejiendo la Web*.
</q>
```
<q>
Yo llegué en el momento justo, interesado, cuando el hipertexto e 
Internet habían visto ya la luz. La tarea que me correspondía era hacer que casaran. <br><br>
Tim Berners-Lee, *Tejiendo la Web*.
</q>
<br><br>

	<strong>Información!</strong><br>    
	<span style=font-style:italic;margin-left:12px;display:block;>
        Indentación es un anglicismo (de la palabra inglesa indentation) de uso común en informática; no es un término reconocido por la Real Academia Española. La Real Academia recomienda utilizar "sangrado". Este término significa mover un bloque de texto hacia la derecha insertando espacios o tabuladores, para así separarlo del margen izquierdo y mejor distinguirlo del texto adyacente; en el ámbito de la imprenta, este concepto siempre se ha denominado sangrado o sangría.
	</span>

### Cita anidada

Una cita anidada es simplemente una "cita dentro de otra cita". Generar citas anidadas en Markdown requiere solo agregar un carácter "__>__" adicional antes de la primera línea de la cita anidada, por ejemplo: 

```markdown
> Cita 1
>> Cita 2 
>>> Cita 3
```
> Cita 1
>> Cita 2 
>>> Cita 3


Entonces, de acuerdo al ejemplo anterior, el úmero de caracteres "__>__" determina el nivel de sangrado o de *profundidad* de la cita anidada. 

Si se requieren más de un salto de línea o líneas en blanco dentro de las citas, solo se debe agregar los caracteres "__>__" en la línea en blanco de acuerdo a su nivel de profundidad:

```markdown
> Cita 1
>
>> Cita 2 
>>
>>> Cita 3
```
> Cita 1
>
>> Cita 2 
>>
>>> Cita 3

En el caso de la sintaxis de HTML se tienen que anidar las etiquetas de cita para generar bloques de cita anidadas.  Por ejemplo:

```html
<blockquote>
    Cita 1
    <blockquote>
        Cita 2
        <blockquote>    
            Cita 3
        </blockquote> 
    </blockquote>   
</blockquote>
```
<blockquote>
    Cita 1
    <blockquote>
        Cita 2
        <blockquote>    
            Cita 3
        </blockquote> 
    </blockquote>   
</blockquote>

## Listas

Otro de los elementos muy utilizados al crear documentos son las listas. En realidad son muy comunes al grado que lenguajes de marcado ligero como Markdown incluyen una sitaxis para poder crealas y, por supuesto, también es posible en lenagujes de marcado como HTML. En esta sección se abordará cómo organizar elementos a través de listas *ordenadas*, *no-ordenadas*, *listas de definición* y *listas de tareas*.

### Listas ordenadas

Tanto Markdown como HTML se pueden crear *listas ordenadas*. Esto es, los elementos que estén enlistados serán enumerados u ordenados alfabéticamente según sea el caso.

En Markdown para crear listas ordenadas se usa una base numérica seguida de un punto, un espacio y el texto a enlistar:

```markdown
1. Elemento 1
2. Elemento 2
3. Elemento 3
```
1. Elemento 1
2. Elemento 2
3. Elemento 3

El ejemplo anterior no es muy ilustrativo pues no parece haber ningún efecto con la sintaxis de Markdown. Sin embargo, la numeración no tiene que ser precisamente continua porque automaticamente Markdown ajustará la secuencia y orden numérico de la lista: 

```markdown
1. Elemento 1
3. Elemento 2
5. Elemento 3
```
9. Elemento 1
3. Elemento 2
5. Elemento 3

Si la lista consta de sublistas o listas anidadas, se pueden crear tabulando (indentando) los elementos que corresponden a la sublista con la tecla <kbd style="font-weight:bold;">Tab &#8633;</kbd>. En el caso de los elementos que sean parte de una sublista, estos serán etiquetados de forma distinta (y en orden) de acuerdo al nivel de profundidad de las listas anidadas:

```markdown
1. Elemento 1
2. Elemento 2
    3. Sub-elemento 1
    4. Sub-elemento 2
        1. Sub-elemento 1
        7. Sub-elemento 2
```
1. Elemento 1
2. Elemento 2
    3. Sub-elemento 1
    4. Sub-elemento 2
        1. Sub-elemento 1
        7. Sub-elemento 2

Cabe resaltar que si se quiere continuar el orden dentro de las listas anidadas, se tiene que seguir la numeración en orden ascendente para mantener el orden etiquetado por Markdown. Para reiniciar el orden etiqueta de las sublistas se tiene que empezar a enumerar desde el inicio:

**Lista con orden continuado**
```markdown
1. Elemento 1
2. Elemento 2
    3. Sub-elemento 1
    4. Sub-elemento 2
```
1. Elemento 1
2. Elemento 2
    3. Sub-elemento 1
    4. Sub-elemento 2

**Lista con orden reiniciado**
```markdown
1. Elemento 1
2. Elemento 2
    1. Sub-elemento 1
    2. Sub-elemento 2
```
1. Elemento 1
2. Elemento 2
    1. Sub-elemento 1
    2. Sub-elemento 2

Por otro lado, para enlistar de forma ordenada en HTML se usan dos etiquetas: "<tt style="font-weight:bold;">&lt;ol&gt;</tt>" y "<tt style="font-weight:bold;">&lt;li&gt;</tt>".

La etiqueta "<tt style="font-weight:bold;">&lt;ol&gt;</tt>" define el elemento de listas ordenadas; mientras que la etiqueta "<tt style="font-weight:bold;">&lt;li&gt;</tt>" define los elemetos que pertenecen a la lista ordenada:

```html
<ol>
    <li>Primer elemento</li>
    <li>Segundo elemento</li>
    <li>Tercer elemento</li>
</ol>
```
<ol>
    <li>Primer elemento</li>
    <li>Segundo elemento</li>
    <li>Tercer elemento</li>
</ol>

Para anidar listas se tienen que anidar las etiquetas anteriores con sus respectivos elementos como se muestra en el siguiente caso:

```html
<ol>
    <li>Primer elemento</li>
    <li>Segundo elemento</li>
    <li>Tercer elemento</li>
    <ol>
        <li>Primer subelemento</li>
        <li>Segundo subelemento</li>
    </ol>
</ol>
```
<ol>
    <li>Primer elemento</li>
    <li>Segundo elemento</li>
    <li>Tercer elemento</li>
    <ol>
        <li>Primer subelemento</li>
        <li>Segundo subelemento</li>
    </ol>
</ol>

A diferencia de Markdown, en HTML se puede personalizar cómo se enlistan los elementos usando tres atributos en la etiqueta "<tt style="font-weight:bold;">&lt;ol&gt;</tt>":

1. <tt><b>reversed</b></tt>: Especifica que el orden de la lista debe invertirse.

2. <tt><b>start</b></tt>: Especifica a partir de qué valor inicia la lista.

3. <tt><b>type</b></tt>: Especifica el tipo de etiqueta o marcador de la lista. Las opciones son:
    - 1
    - A
    - a
    - I
    - i

Por ejemplo, para mantener el orden en sublistas se usa el atributo "<tt><b>start</b></tt>" para continuar en el valor que se especifica:

```html
<ol>
    <li>Primer elemento</li>
    <li>Segundo elemento</li>
    <li>Tercer elemento</li>
    <ol start='4'>
        <li>Primer elemento</li>
        <li>Segundo elemento</li>
        <ol>
            <li>Primer elemento</li>
            <li>Segundo elemento</li>
        </ol>
    </ol>
</ol>
```
<ol>
    <li>Primer elemento</li>
    <li>Segundo elemento</li>
    <li>Tercer elemento</li>
    <ol start='4'>
        <li>Primer elemento</li>
        <li>Segundo elemento</li>
        <ol>
            <li>Primer elemento</li>
            <li>Segundo elemento</li>
        </ol>
    </ol>
</ol>

### Listas no-ordenadas

Las listas no-ordenadas básicamente son elementos enlistados pero sin orden numérico o alfabético. En su lugar se usan *viñetas*, tales como, círculos o cuadros. 

En Markdown para crear listas no-ordenadas se usan los signos: +, \*, o -.

```markdown
- Elemento 1
+ Elemento 2
* Elemento 3
    - Sub-elemento 1
    - Sub-elemento 2
        + Sub-elemento 1
        + Sub-elemento 2
    - Sub-elemento 3
- Elemento 4        
```
- Elemento 1
+ Elemento 2
* Elemento 3
    - Sub-elemento 1
    - Sub-elemento 2
        + Sub-elemento 1
        + Sub-elemento 2
    - Sub-elemento 3
- Elemento 4

No importa qué signos se usen para enlistar los elementos, las marcas que enlistan los elementos se asignan de acuerdo al nivel de profundidad de las listas para distinguirlas unas de otras. Y de igual manera que las listas ordenadas, las sublistas se crean simplemente tabulando o indentando los elementos de las sublistas. 

Por otra parte, usar HTML crear listas no-ordenadas requiere de la etiqueta "<tt style="font-weight:bold;">&lt;ul&gt;</tt>" para generar la estructura de la lista no ordenada. Para agregar los elementos a una lista no-ordenada se emplea la etiqueta "<tt style="font-weight:bold;">&lt;li&gt;</tt>":

```html
<ul>
    <li>Elemento 1</li>
    <li>Elemento 2</li>
    <li>Elemento 3</li>    
</ul>
```
<ul>
    <li>Elemento 1</li>
    <li>Elemento 2</li>
    <li>Elemento 3</li>    
</ul>

Y para crear listas anidadas se usa la misma lógica que con listas ordenadas en HTML:

```html
<ul>
    <li>Elemento 1</li>
    <li>Elemento 2</li>
    <li>Elemento 3</li>  
    <ul>
        <li>Sub-elemento 1</li>
        <li>Sub-elemento 2</li>
        <ul>
            <li>Sub-elemento 1</li>
            <li>Sub-elemento 2</li>
        </ul>
         <li>Sub-elemento 3</li> 
    </ul>
</ul>
```
<ul>
    <li>Elemento 1</li>
    <li>Elemento 2</li>
    <li>Elemento 3</li>  
    <ul>
        <li>Sub-elemento 1</li>
        <li>Sub-elemento 2</li>
        <ul>
            <li>Sub-elemento 1</li>
            <li>Sub-elemento 2</li>
        </ul>
         <li>Sub-elemento 3</li> 
    </ul>
</ul>

Las viñetas asignadas para enlistar los elementos son las mismas que se asignan usando la sintaxis de Markdown. 

### Listas de definición 

Las *listas de definición* se componenen de un "término" y una "definición". No todos los [procesadores de marcado](#Markdown-vs-HTML "clic aquí") soportan esta característica y concretamente Markdown no lo soporta. Sin embargo, mediante HTML es posible crear este tipo de listas usando las etiquetas: <tt style="font-weight:bold;">&lt;dl&gt;</tt>, <tt style="font-weight:bold;">&lt;dt&gt;</tt> y <tt style="font-weight:bold;">&lt;dd&gt;</tt>.

La etiqueta "<tt style="font-weight:bold;">&lt;dl&gt;</tt>" define la estructura de la lista, "<tt style="font-weight:bold;">&lt;dt&gt;</tt>" define el termino y "<tt style="font-weight:bold;">&lt;dd&gt;</tt>" la descripción. El siguiente ejemplo clarifica cómo emplear estas etiquetas:

```html
<dl>
  <dt>Café</dt>
      <dd>Bebida caliente color negro</dd>
  <dt>Leche</dt>
      <dd>Bebida fría color blanco</dd>
</dl>
```
<dl>
  <dt>Café</dt>
      <dd>Bebida caliente color negro</dd>
  <dt>Leche</dt>
      <dd>Bebida fría color blanco</dd>
</dl>

Ahora, si bien es cierto que el procesador de marcado Markdown no permite este tipo de listas, se puede "emular" la estructura de estas listas usando la sintaxis de listas ordenadas y no-ordenadas a la vez:

```markdown
1. Café
    - Bebida caliente color negro
2. Leche
    - Bebida fría color blanco
```
1. Café
    - Bebida caliente color negro
2. Leche
    - Bebida fría color blanco

Aunque por supuesto no produce el mismo tipo de formato es una forma alterna de poder emular las listas de definición usando Markdown. 

### Listas de tareas 

Las *listas de tareas* son listas con una sintaxis especial que permite crear cajas de verificación (checkboxes) para los elementos de una lista. 

Crear listas de tareas requiere emplear guiones medios para enlistar los elementos, un par de chorchetes y una "__x__" para agregar un marca de selección a la caja de verificación:

```markdown
- [x] Tarea realizada 
- [ ] Tarea pendiente
```
- [x] Tarea realizada 
- [ ] Tarea pendiente

Es importante mencionar que las cajas de verificación vacía requiere un espacio en blanco entre los corchetes. Sino se deja un espacio la sintaxis no se aplica correctamente:

```markdown
- [x] Tarea realizada 
- [] Tarea pendiente
```
- [x] Tarea realizada 
- [] Tarea pendiente

## Tablas

Crear tablas en Markdown resulta ser algo más sencillo en contraste con HTML. Pero aunque es más fácil crear tablas en Markdown, HTML tiene mayores ventajas para el diseño y opciones de tabla. 

### Tablas en Markdown 

Las tablas en Markdown requiere solo dos signos: guiones y barras verticales (plecas).

Las plecas se utilizan para definir y separar celdas de datos mientras que los guiones definen la separación entre los encabezados y los datos de la tabla además de definir las columnas (el número de guiones es igual al número de columnas):

```markdown
A|B|C
-|-|-
1|2|3
4|5|6
```
A|B|C
-|-|-
1|2|3
4|5|6
<br>

En el ejemplo anterior, las letras "A", "B" y "C" son los nombres de los encabezados los cuales son separados por barras verticales. Luego, se agregan guiones y plecas para separar los encabezados de los datos. Finalmente se agregan las filas de datos donde cada celda es separada por plecas.

Es necesario tomar en cuenta que a longitud de las filas tiene que ser igual a la longitud de encabezados. Si se requiere agregar columnas adicionales se tiene que agregar plecas y guiones adicionales para agregar las columnas necesarias. De no hacerlo, se pierde elementos de la tabla:

```markdown
A|B|C| |
-|-|-|-
1|2|3|
4|5|6|7
```
A|B|C| |
-|-|-|-
1|2|3|
4|5|6|7

```markdown
A|B|C| 
-|-|-|
1|2|3|
4|5|6|7
```
A|B|C| 
-|-|-|
1|2|3|
4|5|6|7
<br>

Y para agregar filas adicionales que no contengan datos se deben agregar las plecas aunque no contengan datos:

```markdown
A|B|C| |
-|-|-|-
1|2|3|
| | |
7|8|9|10 
```
A|B|C| |
-|-|-|-
1|2|3|
| | |
7|8|9|10
<br>

Otra carcaterística adicional de las tablas en Markdown es que se puede alinear los encabezados y el contenido de las celdas. Se puede alinear "hacia la izquierda", "hacia la derecha" o "centrarlo".

Para definir la alineación se usan dos puntos (__:__) como sintaxis:

- Alinear a la izquierda: <span style='font-size:25px'>:---</span>

- Alinear a la derecha: <span style='font-size:25px'>---:</span>

- Centrado: <span style='font-size:25px'>:---:</span>

En el siguiente ejemplo se muestra una tabla con los tres tipos de alineación:
    
```markdown
Izquierda|Centrado|Derecha
:--------|:------:|------:
<<|<<>>|>>
1|1|1
12|11|21
123|121|321
1234|12321|4321
```
Izquierda|Centrado|Derecha
:--------|:------:|------:
<<|<<>>|>>
1|1|1
12|11|21
123|121|321
1234|12321|4321
<br>

Todas las tablas siempre se quedan centradas una vez que se procesa la sintaxis en la celda de markdown.

### Tablas en HTML
    
Crear tablasen HTML es un poco más laborioso que su contraparte Markdown. Sin embargo, la cantidades de opciones para personalizar el diseño de una tabla y sun funcionalidad se extendiende mucho más allá de las posibilidades de Markdown y en algunos casos es más conveniente para cierto tipo de tablas que requieran mayor detalle. Crear y personalizar tablas en HTML es un tema bastante amplio por lo que en esta sección solo se abordarán los detalles generales de las mismas. 
    
Para crear tablas básicas en HTML se requieren varias etiquetas para crear la estructura del elemento HTML. La principal etiqueta que define una tabla es "<tt style="font-weight:bold;">table</tt>". En esta etiqueta se contienen otras etiquetas para dar forma a la tabla:

- <tt><b>&lt;tr&gt;&lt;/tr&gt;</b></tt>: Define las filas de la tabla.

- <tt><b>&lt;th&gt;&lt;/th&gt;</b></tt>: Define los encabezados de la tabla. 

- <tt><b>&lt;td&gt;&lt;/td&gt;</b></tt>: Define las celdas para los datos de la tabla. 

Cada una etiqueta "<tt style="font-weight:bold;">&lt;tr&gt;</tt>" define una fila en la tabla y, dentro de ella, se definen los encabezados o celdas de datos dependiendo la etiqueta usada. Por ejemplo, en el siguiente caso se crea una tabla que contiene una sola fila para el encabezado y dos filas para los datos:

```html
<table>
  <tr>
    <th>Mes</th>
    <th>Ingresos</th>
  </tr>
  <tr>
    <td>Enero</td>
    <td>&#36;100</td>
  </tr>
  <tr>
    <td>Febrero</td>
    <td>&#36;80</td>
  </tr>
</table>
```

En el caso del pie de tabla se ha agregado además el atributo <tt style="font-weight:bold;">colspan</tt> con el cual se pueden ajustar la distribución del pie de página a lo ancho de la tabla; es decir, se puede ajustar el pie de tabla solo a lo ancho de una columna o distribuirse entre varias columnas para una mejor presentación. Además, para que aparezca centrado el pie de tabla se usa el atributo <tt style="font-weight:bold;">align</tt> en la etiqueta <tt style="font-weight:bold;">tfoot</tt> donde se establece el valor correspondiente para distribuir de forma correcta el pie de tabla. 

Hasta aquí hemos visto algunos de los elementos principales de una tabla HTML para poder personalizar y diseñar su contenido de forma que sea más legible o presentable al lector. Por supuesto hay muchos otros elementos que se pueden añadir pero eso implica un mayor trabajo para escribir el lenguaje de marcado.

En aras de la simpllicidad se recomienda emplear la sintaxis de Markdown sobre la de HTML para generar tablas sencillas. Pero si la tabla requiere una mejor presentación de resultados en un notebook entonces se puede emplear la sintaxis de HTML. 

En Internet existen muchos recursos orientados en la creación de tablas en HTML. En los siguientes enlaces se puede encontrar una guía sobre varios aspectos de tablas HTML tanto básicos como avanzados así como un sitio que genera código HTML llenando solo los campos que se necesitan de la tabla:

1. [Tutorial de tablas HTML](http://www.manualweb.net/html/tablas-html/ "clic aquí")

2. [Generador de tablas HTML](https://www.tablesgenerator.com/html_tables)

## Enlaces

Otra característica muy útil en un documeno electrónico es incluir enlaces (links) para hacer referencia a algún sitio externo, tal como ua página web. También en algunas ocasiones es necesario hacer referencias a ciertos recursos/ secciones u otros elementos en un notebook por lo que también es necesario crear enlaces. 

Tanto Markdown como HTML proveen de una sintaxis para crear enlaces tanto a referencias externas como internas a un notebook.

### Enlaces en Markdown

Existen diversas formas de crear enlaces en Markdown. Básicamente existen dos tipos de enlaces: enlaces externos y enlaces internos.

Los enlaces externos permiten crear referencias a recursos fuera de un documento electrónico o notebook tales como un sitio web o para ejecutar el motor de correo electrónico predeterminado. Por su lado, los enlaces internos permiten crear referencias a recursos que se encuentran dentro de un documento electrónico los cuales pueden ser referencias a tablas, secciones e incluso imágenes. 

Hay diversas formas de crear y personalizar enlaces de acuerdo a los requerimentos del mismo:

- __Enlaces automáticos__

La forma más inmediata de crear un enlace en Markdown es simplemente pegar el enlace en el documento y Markdown automáticamente lo convertirá en un enlace, es decir, que al hacer clic sobre el enlace automáticamente abrirá el recurso a la referencia que se haya colocado:

```markdown
https://jupyter.org/
```
> https://jupyter.org/

De esa forma y sin necesidad de indicar a Markdown que se trata de un  enlace, automáticamente el recurso adopta la función de una [<span style="text-decoration:underline;">URL</span>](https://es.wikipedia.org/wiki/Localizador_de_recursos_uniforme "clic aquí") que se formatea en color azul (caracteístico de enlaces) y submarcado. Por supuesto, el enlace tiene que ser correctamente escrito para que funcione. Si solo se desea que el enlace aparezca como texto y no como un recurso *clickable*, entonces, simplemente se tiene que poner el enlace entre dos comillas invertidas: 

```markdown
`https://jupyter.org/`
```
> `https://jupyter.org/`

- __Enlaces descriptivos__

En algunas ocasiones se requerirá que el enlace no se vea como una URL sino como un texto que haga sentido a la referencia del enlace. Para crear enlaces que contengan texto en lugar del enlace crudo se usa una sintaxis basada en corchetes. Dentro de los corchetes se coloca el texto a mostrar; mientras que entre los paréntesis se escribe como tal el enlace al sitio que se haga referencia:

```markdown
[Página de Jupyter.org](https://jupyter.org/)
```
> [Página de Jupyter.org](https://jupyter.org/)

A este tipo de enlaces se les denomina descriptivos o personalizados.

- __Enlaces a correo electrónico__

También está la posibilidad de enlazar con algún gestor de correo electrónico con la dirección de correo electrónico del destinatario especiicado. Ls sintaxis es similar a la de enlaces personalizados con la salvedad de agregar el término "<tt><b>mailto:</b></tt>" antes de la dirección de correo electrónico deseado:

```markdown
[Correo electrónico](mailto:user@mail.to)
```
> [Correo electrónico](mailto:user@mail.to)

Al hacer clic sobre el enlace se abrirá, por defecto, el gestor de correo predeterminado del sistema (ej. Outlook en caso de Windows). Cabe destacar que la dirección de correo electrónico debe escribirse sin espacios después del término "<tt><b>mailto:</b></tt>", de lo contrario, la sintaxis no funcionará. 

- __Enlaces descriptivos con título__

También es posible mostrar un "título" en un enlace al pasar el puntero del mouse sobre él sin necesidad de hacer clic. Esto se logra agregando el texto a mostrar entre comillas después del enlace de referencia y separado por un espacio:

```markdown
[Enlace con título](https://www.algo.com/ "Mi enlace")
```
> [Enlace con título](https://www.algo.com/ "Mi enlace")

Cuando se crea el enlace, al pasar el puntero del mouse sobre el enlace, se mostrará el texto que haya sido colocado entre comillas. 

- __Enlaces referenciados__

Podemos referenciar un enlace sin necesidad de incluir el recurso directamente entre paréntesis. En su lugar se usan corchetes y en su interior se agrega un "id" que hace referencia a un enlace. Es como crear un enlace para referenciar a otro enlace:

```markdown
[Enlace referido][id_enlace]

[id_enlace]: https://www.algo.com/
```

> [Enlace referido][id_enlace]

[id_enlace]: https://www.algo.com/

También es posible definir un enlace referenciado sin usar un "id". Se pueden crear sin agregar un identificador entre los corchetes o incluso se pueden omitir, la referencia se crea con el nombre descriptivo del enlace refenciado:

```markdown
[enlace 1][]

[enlace 1]: https://www.algo.com/ "algo.com"

[enlace 2]

[enlace 2]: https://www.google.com/ "google.com"
```
> [enlace 1][]

[enlace 1]: https://www.algo.com/ "algo.com"

> [enlace 2]

[enlace 2]: https://www.google.com/ "google.com"

- __Enlaces a otros archivos__
    
También es posible crear vínculos a archivos indicando el nombre y ruta del mismo. Por ejemplo, en el directorio "*/Recursos/Archivos*" hay tres tipos de archivos los cuales pueden ser ejecutados mediante los siguientes enlaces:

```markdown
[Archivo de jupyter](Recursos/Archivos/test.ipynb "clic para abrir")

[Archivo de markdown](Recursos/Archivos/test.md "clic para abrir")

[Archivo de texto](Recursos/Archivos/test.txt "clic para abrir")
```
> [Archivo de jupyter](Recursos/Archivos/test.ipynb "clic para abrir")
>
> [Archivo de markdown](Recursos/Archivos/test.md "clic para abrir")
> 
> [Archivo de texto](Recursos/Archivos/test.txt "clic para abrir") 

En estos tres enlaces se especifica la ruta del directorio, nombre de archivo y extensión del mismo para poder abrir el archivo referido. 

Es imprescindible escribir correctamente la ruta del directorio, el nombre y extensión del archivo para que el enlace funcione correctamente. 

<span style="font-size:20px;">Enlaces internos</span>

Hasta ahora se ha visto las distintas formas de crear enlaces a recursos externos, pero también hay formas de crear enlaces para referenciar secciones, tablas, imágenes u otro tipo de elementos dentro de un notebook. 

Crear enlaces a recursos internos usa la misma sintaxis para enlaces externos salvo por algunas diferencias. Además, dependiendo del recurso que se requiere referenciar se tendrá que emplear la URL adecuado.

- __Enlaces a secciones__

Un enlace muy útil al escribir un documento electrónico es aquel para hacer referencia a una sección específica del mismo. Por ejemplo, si se requiere crear un enlace para referenciar a la sección "**Creando u nnotebook**" se tiene que emplear la siguiente sintaxis:

```markdown
[Creando un notebook](#Creando-un-notebook)
```
> [Creando un notebook](#Creando-un-notebook)

Se debe notar que entre los paréntesis en lugar de escribir una URL, se escribe el nombre de la sección tal cual aparece ya que Markdown distingue mayúsculas y minúsculas. Además se debe anteponer siempre una almohadilla (__#__) antes del nombre de la sección y, si hay espacios en el nombre de la sección, estos deben ser sustituidos por un guión medio. De omitir la almohadilla o no sustituir los espacios en blanco por guiones, el enlace no funcionará:

```markdown
[enlace inválido](Lenguajes-de-Marcado)

[enlace inválido](#Lenguajes de Marcado)
```
> [enlace inválido](Lenguajes-de-Marcado)

> [enlace inválido](#Lenguajes de Marcado)

Cabe mencionar que estos enlaces solo funcionan para secciones creadas usando la sintaxis de encabezados vía Markdown o HTML.

- __Enlaces a tablas__

Las secciones no son lo único a lo cual se puede hacer referencia. Se pueden crear enlaces para referenciar tablas en un notebook:

```markdown
[Tabla 1](#T1 "Formatos de énfasis")
```
> [Tabla 1](#T1 "Formatos de énfasis")

En este ejemplo se hace referencia a la primera tabla de este notebook. Esta tabla puede ser referenciada gracias a un *identificador* definido en el atributo global "<tt style="font-weight:bold;">id</tt>" al crear una tabla mediante HTML. Es este caso el identificador de la tabla 1 es "__T1__", pero podría ser cualquier otro nombre. Más adelante en "[Enlaces en HTML](#Enlaces-en-HTML)" se definirá como emplear el atributo "<tt style="font-weight:bold;">id</tt>".

Es importante señalar que solo se puede hacer referencia a tablas que sean creadas con sintaxis HTML.

- __Enlaces a imágenes__

En un notebook es posible insertar imágenes así como también referenciarlas mediante un enlace en cualquier parte de un notebook. Las imágenes insertadas mediante HTML pueden emplear el atributo "<tt style="font-weight:bold;">id</tt>" con el cual se puede establecer una referencia o identificador que enlace hacia la imagen. 

Por ejemplo, el siguiente enlace puede hacer referencia a la imagen de la sección "[<span style="text-decoration:underline;">Jupyter Server</span>](#Jupyter-Server "clic aquí")":

```markdown
[Imagen Jupyter Server](#fig_ref)
```
> [Imagen Jupyter Server](#fig_ref)

De nuevo, la imagen insertada mediante código HTML permite definir el atributo "<tt style="font-weight:bold;">id</tt>" en el cual se estableció el identificador "<tt style="font-weight:bold;">fig_ref</tt>" aunque podría ser cualquier otro identificador.  

Nuevamente, referenciar imágenes solo es posible usando sintaxis HTML para insertar imágenes.

### Enlaces en HTML
    
Para crear enlaces en HTML se usa la etiqueta "<tt style="font-weight:bold;">&lt;a&gt;</tt>" (del inglés *anchor*) misma que acepta el atributo "<tt style="font-weight:bold;">href</tt>" para indicar la URL a la cual se hará referencia:

```html
<a href='https://jupyter.org/'>Jupyter.org</a>
```
> <a href='https://jupyter.org/'>Jupyter.org</a>

En este ejemplo, en el atributo "<tt style="font-weight:bold;">href</tt>" se establece el enlace al cual se hará referencia y entre las etiquetas de apertura y cierre se especifica el nombre descriptivo para el enlace.

Si además se quiere mostrar un título personalizado al pasar el puntero del mouse sobre el enlace se debe emplear el atributo "<tt style="font-weight:bold;">title</tt>":

```html
<a href='https://jupyter.org/' title='Enlace a Jupyter org'>Jupyter.org</a>
```
> <a href='https://jupyter.org/' title='Enlace a Jupyter org'>Jupyter.org</a>

Si en lugar de hacer referencia a un sitio web se requiere ejecutar el motor de email, solo se debe anteponer el prefijo "<tt style="font-weight:bold;">mailto:</tt>" seguido de la dirección del correo electrónico de destino:

```html
<a href="mailto:user@mail.to">Correo electrónico</a>
```
> <a href="mailto:user@mail.to">Correo electrónico</a>

También se puede autoreferenciar un enlace HTML usando el atributo "<tt style="font-weight:bold;">id</tt>" donde se especifica el identificador del elemento enlace:

```html
<a id='E1' href='https://jupyter.org/'>Jupyter.org</a>

<a href='#E1'>Referencia a enlace 1</a>
```
> <a id='E1' href='https://jupyter.org/'>Jupyter.org</a>
>
> <a href='#E1'>Referencia a enlace 1</a>

Es importante notar que se debe anteponer la almohadilla (__#__) seguido del nombre definido en el atributo "<tt style="font-weight:bold;">id</tt>" del primer enlace en el segundo enlace que hace referencia al primero. 

Si se requiere crear enlaces para abrir o ejecutar algún archivo en específico solo se debe especificar la ruta y nombre de archivo en el atributo "<tt style="font-weight:bold;">href</tt>":

```html
<a href='Recursos/Archivos/test.ipynb'>notebook</a>

<a href='Recursos/Archivos/test.md'>markdown</a>

<a href='Recursos/Archivos/test.txt'>texto</a>
```
> <a href='Recursos/Archivos/test.ipynb'>notebook</a>
>
> <a href='Recursos/Archivos/test.md'>markdown</a>
>
> <a href='Recursos/Archivos/test.txt'>texto</a>

<span style="font-size:20px;">Enlaces internos</span>

Crear enlaces internos usando HTML también es posible. La idea es la misma que usando sintaxis de Markdown: se antepone una almohadilla antes del identificador o URL de la sección, imágen, tabla o cualquier elemento que se requiere enlazar:

```html
<a href="#toc">Tabla de contenidos</a>

<a href="#T2">Tabla 2</a>

<a href="#toolbar">Imagen: Barra de herramientas</a>

```
> <a href="#toc">Tabla de contenidos</a>
>
> <a href="#T2">Tabla 2</a>
>
> <a href="#toolbar">Imagen: Barra de herramientas</a>

Nuevamente, si la URL contiene espacios, estos deben ser sustituídos por guiones medios para que que el enlace sea funcional.

<span style="font-weight:bold;font-size:15px;">Atributo id</span>

El atributo global "<tt style="font-weight:bold;">id</tt>" es una propiedaden HTML que define un identificador único (ID) el cual no debe repetirse en todo el documento electrónico. Su propósito es identificar el elemento al vincularlo mediante un enlace.  

Por ejemplo, consideremos los siguientes casos:

```html
<table id="tabla">

<img id='imagen'>
    
<h1 id='enacabezado'>

<p id='parrafo'>

<a id='enlace'>

<span id='contenedor'>
```
 
Los casos anteriores son ejemplos de etiquetas en HTML que soportan el atributo "<tt style="font-weight:bold;">id</tt>". En cada una de ellas se define un único identificador que luego puede ser empleado como URL en enlaces creados mediante Markdown o HTML. 

Nótese también que las imágenes, tablas o encabezados no son lo único que se puede vincular. Se pueden crear vínculos hacia párrafos o porciones de texto empleando las etiquetas correctas que soporten el atributo <tt style="font-weight:bold;">id</tt>.

<span style="font-weight:bold;font-size:15px;">Codificación de URL</span>

Un problema con los enlaces HTML es que los caracteres especiales tales como acentos o signos de interrogación tienen que estar codificados a un formato que pueda ser reconocido por HTML. 

Para entender por qué deben estar codificados los caracteres especiales, consideré el siguiente caso:

```html
<a href='#Instalación'>Instalación</a>
```
> <a href='#Instalación'>Instalación</a>

El enlace anterior crea un vínculo a la sección "__Instalación__". La sintaxis del enlace es correcta, pero al hacer clic en el enlace éste no vincula al recurso deseado. El hecho que no funcione es debido a que el nombre de sección tiene una letra acentuada lo cual es un carácter especial que no puede ser interpretado correctamente por HTML debido a su formato.

La **codificación de URL** convierte los caracteres en un formato que se puede transmitir a través de Internet. Las direcciones URL solo se pueden enviar a través de Internet utilizando el conjunto de caracteres [<span style="text-decoration:underline;">ASCII</span>](https://es.wikipedia.org/wiki/ASCII "clic aquí").

Dado que las URL a menudo contienen caracteres fuera del conjunto ASCII, la URL debe convertirse a un formato ASCII válido. La codificación de URL reemplaza los caracteres ASCII no seguros con un signo de porcentaje "__%__" seguido de dos dígitos [<span style="text-decoration:underline;">hexadecimales</span>](https://es.wikipedia.org/wiki/Sistema_hexadecimal "clic aquí"). Por ejemplo, los códigos de los caracteres "__?__", "__/__" y "__=__" son <tt><b>%3F</b></tt>, <tt><b>%2F</b></tt> y <tt><b>%3D</b></tt>, respectivamente.

Además, las direcciones URL no pueden contener espacios. La codificación de URL normalmente reemplaza un espacio con un signo más (__+__), guión medio (__-__) o con <tt style="font-weight:bold;">%20</tt>.

Para entender las codificaciones de URL, considérese el siguiente ejemplo: la codificación de la letra "__A__" (mayúscula) es "<tt><b>%41</b></tt>" el cual representa el número "<tt><b>0x41</b></tt>" en su forma hexadecimal (números en base 16). A su vez, su forma decimal (números en base 10) es el número 65 que en el código ASCII representa la letra A mayúscula. Por lo tanto, la letra "__A__" puede ser codificada como "<tt><b>%41</b></tt>".

Dicho lo anterior, para que el enlace del ejemplo anterior funcione correctamente se tiene que codificar de la siguiente manera:

```html
<a href='#Instalaci%C3%B3n'>Instalación</a>
```
> <a href='#Instalaci%C3%B3n'>Instalación</a>

En Internet hay varios sitios que pueden codificar o decodificar una URL de forma automática como es el caso del sitio: [<span style="text-decoration:underline;">URL Encode/Decode</span>](https://www.url-encode-decode.com/ "clic aquí")

También es posible encontrar referencias de codificaciones de URL y del conjunto de caracteres ASCII como en los siguientes enlaces:

- [<span style="text-decoration:underline;">URL Encoding Reference</span>](https://www.w3schools.com/tags/ref_urlencode.asp "clic aquí")

- [<span style="text-decoration:underline;">ASCII Reference</span>](https://www.w3schools.com/charsets/ref_html_ascii.asp "clic aquí")

## Códigos de muestra

Cuando se trabaja en notebooks a menudo se reportan resultados en celdas markdown que incluyen códigos de ejemplo para que puedan ser reproducidos.   

El propósito de los códigos de ejemplo o de muestra en las celdas markdown es mostrar un ejemplo de código que luego pueda ser ejecutado o copiado en una celda de código.

Markdown proveé una sintaxis para incluir bloques de código que se pueden representar tal y como aparecerían en una celda de código.  

Hay dos formas de mostrar código de ejemplo en la celda markdown: 

1. **Usando comillas simples invertidas (acentos graves)**: Esta sintaxis formateará el código con el tipo de fuente [<span style="text-decoration:underline;">monospace</span>](https://es.wikipedia.org/wiki/Tipo_de_letra_monoespaciado "clic aquí") sobre un fondo resaltado en gris: 

    ```markdown
    `x = 0`
    ```
    
    `x = 0`
     
    Usualmente se emplea este tipo de formato para código de ejemplo que no ocupa más de una línea o que se muestra entre texto.
     
2. **Usando triples comillas invertidas y el nombre del lenguaje (opcional)**: Este formato usa triples comillas invertidas para incluir código de múltiples líneas y que requiere un formato fiel a la sintaxis del lenguaje de programación mostrado en una celda markdown. Markdown soporta una amplia lista de formatos de sintaxis de distintos lenguajes de programación.

        ```
        a, b = 0, 1
        amax = 100 
        L = []
        while True: 
            (a, b) = (b, a + b) 
            if a > amax: 
            break 
            L.append(a) 
            print(L)
        ```

El resultado es un código formateado en fuente *monospaciada*. ## Escape de caracteres

Un problema de usar caracteres especiales en celdas markdown es que algunos caracteres son usados internamente por el procesador de Markdown como funciones para ciertas tareas. Es decir, algunos símbolos especiales tienen un *significado especial* lo que dificulta la realización de algunas tareas de formato. 

Por ejemplo, ya se ha visto que el símbolo almohadila (__#__) es usado por Markdown para indicar el formato de encabezados. Por su lado, los caracteres asterísco (__\*__) y guión bajo (**\_**) son usados para dar énfasis a un texto plano en formato negritas o itálicas.

Para evitar que Markdown interprete los caracteres especiales como funciones internas, se puede emplear una *diagonal inversa o backslash* (\\) para evitar, saltar o escapar del significado especial de caracteres. El siguiente ejemplo lo demuestra:

```markdown
\# Esto escapa de crear un encabezado.
```
> \# Esto escapa de crear un encabezado.

Aquí, la diagonal inversa antes de la almohadila le permite "escapar" de su función para crear encabezados, dejando a la almohadila ser un carácter más del texto. Además, nótese que la diagonal no aparece en el formato de salida pues su función es la de evitar que ciertos caracteres se comporten como caracteres y no como funciones de formato. Es como si la diagonal inversa y la función de los caracteres especiales se cancelaran uno a otro.

## Multimedia

En un notebook también es posible insertar elementos multimedia tales como imágenes, video o audio. Tanto Markdown como HTML proveén la sintaxis necesaria para insertar imágenes, pero solo con HTML es posible incluir video y audio. 

En la práctica es poco probable usar elementos de video/audio en un notebook, pero en esta sección se demostrará cómo incluir esos elementos multimedia. 

### Imagen

En una celda markdown se pueden insertar imágenes ya sea desde un archivo de imágen local o desde un recurso de Internet. 

Para insertar imágenes con Markdown se emplea una sintaxis similar para crear enlaces, pero agregando un signo de exclamación de cierre (__!__) antes de los corchetes y, opcionalmente, un texto decriptivo entre los corchetes:

```markdown
![](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png)

![Imagen de Wikipedia](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png)

```
![](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png)

Como se puede notar esta sintaxis tiene sentido que sea similar para crear un enlace usando Markdown, ya que en realidad se está creando un vínculo a un resurso que se encuentra en Internet, el cual es una imagen. La diferencia la hace el signo de exclamación (carácter especial) que le indica a Markdown que debe insertar la imágen directamente en el notebook. De hecho, si se elimina el signo de exclamación, la sintaxis se convierte en un enlace descriptivo clickable que vincula directamente a la imágen en una nueva pestaña del navegador:

```markdown
[Imagen de Wikipedia](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png)
```
> [Imagen de Wikipedia](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png)

De hecho, esta sintaxis para insertar imágenes comparte las mismas propiedades con la sintaxis para crear un enlace, por lo que también se puede agregar un título que aparezca al colocar el puntero sobre la imágen:

```markdown
![Imagen de Wikipedia](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png "Logo de Jupyter")
```
![Imagen de Wikipedia](https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/250px-Jupyter_logo.svg.png "Logo de Jupyter")

Además es posible insertar una imágen desde un archivo local como en el siguiente caso: 

```markdown
![](Recursos/Imagenes/whale.gif "Imagen GIF")
```
![](Recursos/Imagenes/whale.gif "Imagen GIF")

Nótese que la URL contiene el directorio local de donde es insertada la imágen así como el nombre y extensión de la imágen. 

Entonces, insertar imágenes en Markdown es relativamente fácil y similar a crear enlaces. Sin embargo, la sintaxis está limitada pues no se puede controlar el tamaño de imágen ni su posición ni ninguna otra propiedad. 

En contraste, usando HTML se puede también insertar imágenes pero con la ventaja de que se disponen de varias propiedades para manipular la imágen insertada. La etiqueta usada por HTML para insertar imágenes es "<tt style="font-weight:bold;">&lt;img&gt;</tt>" la cual dispone de varios atributos que permiten controlar cómo insertar una imagen. Cabe mencionar que la etiqueta "<tt style="font-weight:bold;">&lt;img&gt;</tt>" no tiene etiqueta de cierre.

Por defecto, para indicar la URL de una imagen se usa el atributo "<tt style="font-weight:bold;">src</tt>":

```html
<img src='Recursos/Imagenes/jupy_logo.png'>
```
<img src='Recursos/Imagenes/jupy_logo.png'>

Como se puede notar la imagen se carga con su tamaño (dimensiones) original lo cual es un inconveniente cuando se trata de imágenes de dimensiones más grandes. Pero se puede controlar el tamaño de imagen usando los atributos "<tt style="font-weight:bold;">height</tt>" y "<tt style="font-weight:bold;">width</tt>": 

```html
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%">
```
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%">

En este ejemplo se han reducido las dimensiones de la imagen (alto y ancho) a la mitad. También es posible incluso modificar la alineación de la imágen para que se inserte en el centro de la celda en lugar de la izquierda:

```html
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%" align="center">
```
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%" align="center">

Aquí, se usa el atributo "<tt style="font-weight:bold;">align</tt>" para alinear la imagen empleando el valor "<tt style="font-weight:bold;">center</tt>". O también se puede insertar a la derecha:

```html
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%" align="right">
```
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%" align="right">

Además, se puede agregar un título a la imagen para que el texto aparezca cuando se coloca el puntero sobre ella usando el atributo "<tt style="font-weight:bold;">title</tt>":

```html
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%" title="Una imagen local">
```
<img src="Recursos/Imagenes/jupy_logo.png" height="50%" width="50%" title="Una imagen local">

La etiqueta "<tt style="font-weight:bold;">&lt;img&gt;</tt>" también soporta el atributo global "<tt style="font-weight:bold;">style</tt>" con el cual se pueden personalizar varias propiedades del formato de imagen. Por ejemplo, en la siguiente imagen se inserta un marco de 10 pixeles color rojo alrededor de ella:

```html
<img src="Recursos/Imagenes/jupy_logo.png" width="30%" height="auto" style="border:10px solid red">
```
<img src="Recursos/Imagenes/jupy_logo.png" width="30%" height="auto" style="border:10px solid red">

También se puede insertar imágenes con título al pie de imágen usando la etiqueta "<tt style="font-weight:bold;">&lt;figcaption&gt;</tt>". Para ello se tiene que usar, además, la etiqueta "<tt style="font-weight:bold;">&lt;figure&gt;</tt>" en combinación con la etiqueta "<tt style="font-weight:bold;">&lt;img&gt;</tt>":

```html
<figure>
    <img src="Recursos/Imagenes/jupy_logo.png" width="40%" height="40%">
    <figcaption>Figura #: Imagen del logo de Jupyter</figcaption>
</figure>
```

### Video

Otro recurso multimedia que se puede insertar en un notebook son los archivos de video. Markdown no cuenta con una sintaxis para poder insertar videos en un notebook pero a través de HTML es posible lograrlo. 

Es poco usual que se inserte un video en un documento electrónico, más aún si notebook se va a convertir a otro formato (como un PDF), pero puede llegar a ser útil en algunos casos. Si la idea es insertar un video demostrativo para algún propósito puede ser útil agregar un video en una celda markdown. 

Cabe destacar que solo es posible insertar videos desde un directorio local. No es posible insertar un video desde una URL de Internet.

Para insertar un video en un notebook se usa la etiqueta "<tt style="font-weight:bold;">&lt;video&gt;</tt>". Esta etiqueta de hecho comparte varios atributos utilizados en la etiqueta "<tt style="font-weight:bold;">&lt;img&gt;</tt>" para personaliza el formato y la forma en cómo se inserta:

```html
<video src="Recursos/Videos/Jupyterlab.mp4" width="80%" height="80%" title="Video local" controls></video>
```
<center><video src="Recursos/Videos/Jupyterlab.mp4" width="80%" height="80%" title="Video local" controls></video></center><br>

El ejemplo anterior tiene algunos atributos vistos en "<tt style="font-weight:bold;">&lt;img&gt;</tt>", pero además tiene el atributo "<tt style="font-weight:bold;">controls</tt>" que configura el elemento para mostrar los controles de video (Reproducir/volumen). Es recomendable siempre especificar este atributo para mostrar los controles de reproducción del video insertado.

Adicionalmente hay otros atributos interesantes que permiten controlar el comportamiento del video insertado. A continuación se enlistan los más útiles:

- <tt><b>autoplay</b></tt>: Especifica que el video comenzará a reproducirse tan pronto como se cargue en memoria.

- <tt><b>loop</b></tt>: Especifica que el video se reproducirá automáticamente de forma "infinita".

- <tt><b>muted</b></tt>: Especifica que el video se reproduce sin sonido.

- <tt><b>poster</b></tt>: Especifica una imagen que se mostrará mientras se descarga el video, o hasta que el usuario presione el botón de reproducción.

Por otro lado también es posible personalizar el formato del video a través del atributo "<tt style="font-weight:bold;">style</tt>":

```html
<video src="Recursos/Videos/Jupyterlab.mp4" controls width="80%" height="80%" title="Video local" style="border:5px solid #00ffff"></video>
```
<center><video src="Recursos/Videos/Jupyterlab.mp4" controls width="80%" height="80%" title="Video local" style="border:5px solid #00ffff"></video></center><br>

Finalmente, es importante mencionar que los formatos de video actualmente soportados son: *.mp4*, *.webm* y *.ogg*

# LaTeX

$\mathbf{L\!\!^{{}_{\scriptstyle A}} \!\!\!\!\!\;\; T\!_{\displaystyle E} \! X}$ es un tipo de lenguaje de marcado de composición de textos, orientado a la creación de documentos escritos que presenten una alta calidad tipográfica. Por sus características y posibilidades, es usado de forma especialmente intensa en la generación de artículos y libros científicos que incluyen, entre otros elementos, expresiones matemáticas. Es muy utilizado para la composición de artículos académicos, tesis y libros técnicos, dado que la calidad tipográfica de los documentos realizados en LaTeX, se considera adecuada a las necesidades de una editorial científica de primera línea, muchas de las cuales ya lo emplean.

LaTeX está formado por un gran conjunto de macros de [<span style="text-decoration:underline;">TeX</span>](https://es.wikipedia.org/wiki/TeX  "clic aquí"), escrito por [<span style="text-decoration:underline;">Leslie Lamport</span>](https://es.wikipedia.org/wiki/Leslie_Lamport "clic aquí") en 1984, con la intención de facilitar el uso del lenguaje de composición tipográfica, ${\displaystyle \mathbf {T\!_{\displaystyle E}\!X}}$, creado por [<span style="text-decoration:underline;">Donald Knuth</span>](https://es.wikipedia.org/wiki/Donald_Knuth "clic aquí").

Las celdas markdown también puede soportar la sintaxis de marcado de LaTeX para formatear ecuaciones matemáticas. Las expresiones matemáticas son renderizadas gracias a la librería [<span style="text-decoration:underline;">MathJax</span>](https://www.mathjax.org "clic aquí") de JavaScript.

El propósito de este documento no es ahondar en los detalles de sintaxis de LaTeX. Únicamente se abordará la sintaxis general y básica de LaTeX para insertar ecuaciones matemáticas sencillas. Del lector dependerá investigar en otras referencias para insertar expresiones matemáticas más complejas.

## Sintaxis de LaTeX

Para insertar expresiones o símbolos matemáticos en celdas markdown, se usa el signo dolar "&#36;". Este mismo operador se emplea en otros software que soporten el lenguaje LaTeX. 

Por ejemplo, para insertar y renderizar una expresión matemática basta con ingresar la expresión deseada entre dos signos de dolar: 

```markdown
$y = x^2$
```
> $y = x^2$

Mediante esta sintaxis las expresiones matemáticas se insertan y renderizan en una celda markdown con alineación a la izquierda. También es posible centrar una expresión matemática usando dos veces el signo dolar: 

```markdown
$$y = x^2$$
```
$$y = x^2$$

Otra forma alternativa de crear una expresión matemática es usando los delimitadores <samp>begin{equation}</samp> y <samp>end{equation}</samp>:

```markdown
\begin{equation}
E=mc^2
\end{equation}
```
\begin{equation}
E=mc^2
\end{equation}

Con estos delimitadores la expresión matemática queda centrada automáticamente.

También se puede etiquetar expresiones matemáticas con el delimitador "<samp>\tag{}</samp>": 

```markdown
\begin{equation}
\tag{1}
\frac{dN}{dt} = \dot{N} = rN
\end{equation}
```
\begin{equation}
\tag{1}
\frac{dN}{dt} = \dot{N} = rN
\end{equation}

En este ejemplo se etiqueta la expresión matemática con el número "1" aunque en realidad se puede etiquetar con cualquier otra referencia. 

Por otro lado, es posible crear referencias o vínculos a una expresión matemática usando el delimitador "<samp>\label{}</samp>". Este delimitador en realidad es más general, pero para referenciar expresiones matemáticas específicamente se usa el prefijo "<samp>eq:</samp>" y un identificador único para referenciar la ecuación:

```markdown
\begin{equation}
\label{eq:eq1}
\tag{1}
\frac{dN}{dt} = \dot{N} = rN
\end{equation}
```
\begin{equation}
\label{eq:eq1}
\tag{1}
\frac{dN}{dt} = \dot{N} = rN
\end{equation}

Luego, para crear un vínculo a la referencia de la expresión matemática se usa delimitador "<samp>eqref{}</samp>" donde también se usa el prefijo "<samp>eq:</samp>" y el identificador asociado a la expresión matemática referenciada:

```markdown
$\eqref{eq:eq1}$
```
> $\eqref{eq:eq1}$

De esa forma se crea un enlace de referencia hacia la expresión matemática que se requiera enlazar al hacer clic sobre el identificador. 

Para una completa referencia de uso de LaTeX visite la siguiente página: [<span style="text-decoration:underline;">Manual de LaTeX Completo</span>](https://es.wikibooks.org/wiki/Manual_de_LaTeX/Texto_completo "clic aquí")
