# El Proyecto Jupyter
<hr width=240px size=2, align="left">
<hr width=190px size=2, align="left">


[Jupyter Proyect](https://jupyter.org) surge en 2014 como una evolución del proyecto [IPython (Interactive Python)](https://ipython.org/), una potente consola vitaminada para Python. Sin embargo, Jupyter es mucho más ambicioso que IPython, se pretende construir una plataforma agnóstica del lenguaje que ofrezca a los científicos un conjunto de potentes herramientas para trabajar con datos, visualizarlos y poder compartir los resultados.

El proyecto fue iniciado por [Fernando Pérez](https://bids.berkeley.edu/people/fernando-perez), un científico y profesor de la Universidad de California Berkeley. Sin embargo, el proyecto es software libre y está basado en componentes libres, por lo que está apoyado, desarrollado y mantenido por una amplia comunidad.

El origen de su nombre es un homenaje a [Galileo Galilei](https://es.wikipedia.org/wiki/Galileo_Galilei), al que se considera autor del primer "paper" científico de astronomía de la era moderna en 1610, en el cual describe sus observaciones astronómicas a través de un telescopio de las lunas de Júpiter. Galileo demostró que la Tierra orbita el Sol de la misma manera que las lunas de Júpiter orbitan dicho planeta.

La adopción y el éxito de Jupyter ha sido muy grande, de tal modo que se está convirtiendo en un standard para el trabajo de los científicos de datos. Hoy en día la herramienta está integrada algunas de las plataformas y entornos más importantes y extendidas como [Google Cloud](https://cloud.google.com/?hl=es), [Google Colaboratory](https://colab.research.google.com/notebooks/welcome.ipynb), [Microsoft Azure](https://azure.microsoft.com/es-mx/), [AWS (Amazon Web Services)](https://aws.amazon.com/), [IBM Bluemix](https://www.ibm.com/cloud-computing/bluemix/es/node/245), [Databricks Platform](https://databricks.com/), [Rackspace](https://www.rackspace.com/), [Continuum](https://www.continuum.net/), [Jetbrains](https://www.jetbrains.com/).

El Proyecto Jupyter actualmente está implementado en dos plataformas: **Jupyter Notebook** y __JupyterLab__. 

<h2 align=center>Jupyter Notebook</h2>

Es el primer <i>ambiente computacional interactivo basado en la web</i> del Proyecto Jupyter que permite trabajar con notebooks, terminales, editores de texto, explorar archivos entre otros. Su interfaz principal en un explorador de archivos desde el cual se pueden crear y manipular notebook, los cuales se abren cada uno en nuevas pestañas del explorador donde se esté ejecutando la plataforma.


<h2 align=center>JupyterLab</h2>


JupyterLab es la siguiente generación de Jupyter Notebook y que, a diferencia del primero, éste implementa una interface *modular* que ofrece la posibilidad de abrir en una sola ventana de nuestro navegador todos los notebooks, terminales, archivos, editores de texto, visualizaciones, etc. que el usuario desee abrir. Su flexibilidad permite incluso disponer y ordenar pestañas en *grupos* de acuerdo a como el usuario se sienta más cómodo para trabajar. Esto ofrece una experiencia más cercana a como se trabajaría en un [IDE](https://es.wikipedia.org/wiki/Entorno_de_desarrollo_integrado). 

<h2 align=center>Kernels</h2>

Jupyter soporta integración con más de 40 lenguajes de programación en los que podemos escribir el código de nuestro notebooks, por ejemplo: [Python](https://www.python.org/), [Julia](https://julialang.org/), [R](https://www.r-project.org/) , [Scala](https://www.scala-lang.org/) , [Ruby](https://www.ruby-lang.org/es/) o [Go](https://golang.org/). Pero también es fácilmente integrable con herramientas y plataformas de Big Data como [Spark](https://spark.apache.org), lo que permite abstraerse de la complejidad de estas herramientas, aprovehando todo su potencial desde un entorno muy amigable. Esto es posible através de un sistema basado en [Kernel](https://jupyter-client.readthedocs.io/en/stable/kernels.html). Cada Kernel es un motor de ejecución para un lenguaje o plataforma concreta que ejecuta en el servidor. A través de un sistema de colas basado en [ZMQ](https://zeromq.org/) el código que necesita ser ejecutado es enviado al Kernel correspondiente.

Cada Kernel tiene una forma particular de instalación, y tanto la lista de Kernels que se pueden instalar en Jupyter como la descripción de sus implementaciones se encuentran en el siguiente enlace:

> [**Kernels de Jupyter**](https://github.com/jupyter/jupyter/wiki/Jupyter-kernels)

<h2 align=center>Extensiones</h2>

Adicionalmente, tanto Jupyter Notebook como JupyterLab disponen de una serie de *extensiones* que permiten "aumentar" las funciones y productividad de ambas interfaces; sin embargo, los dos proyectos tienen equipos de desarrollo distinto y las extensiones varian ampliamente entre ellos. Es por esto que el procedimiento para instalar y usar las extensiones se debe hacer de acuerdo a la documentación de cada proyecto: 

> [**Extensiones para Jupyter Notebook**](https://github.com/ipython-contrib/jupyter_contrib_nbextensions)

> [**Extensiones para JupyterLab**](https://jupyterlab.readthedocs.io/en/stable/user/extensions.html) 

Finalmente, antes de proceder a instalar cualquiera de los proyectos de Jupyter, es posible probar las funcionalidades que nos ofrece cada uno de los sistemas antes mencionados en notebooks de prueba o propios, esto es posible a través de este [enlace](https://jupyter.org/try "Try Jupyter").

# Notebooks de Jupyter
<hr width=230px size=2, align="left">
<hr width=180px size=2, align="left">

Los documentos de Jupyter son un sistema de documentos electrónicos llamados *notebooks* los cuales se ejecuta a través de una shell interactiva vía web, a la que podemos acceder desde un navegador sin que estos requieran conexión a Internet. La shell está organizada en pequeños bloques, cada bloque puede contener texto arbitrario formateado en **Markdown**, fórmulas matemáticas en **LaTeX**, código en multitud de lenguajes de programación, resultados, gráficos, vídeos, widgets o cualquier elemento multimedia.

Podemos escribir código de programación en estas celdas e ir ejecutándolo paso a paso o todo de golpe, obteniendo todos los resultados parciales. También podemos usar los bloques de texto para documentar el código o añadir las explicaciones oportunas, que pueden contener enlaces, imágenes, vídeos u otros elementos.

Esta serie de piezas de código, notas y resultados se guardan en un notebook, que es un fichero que contiene toda esta información. Uno de los principales objetivos de Jupyter es fomentar y simplificar la compartición de conocimiento y resultados a través de los notebooks. Plataformas como [GitHub](https://github.com/) o [Databricks Community Edition](https://community.cloud.databricks.com/) facilitan esta tarea. De esta manera los notebooks pueden ser fácilmente difundidos y los resultados pueden ser reproducidos y validados en diferentes entornos. Por supuesto esto es muy útil para la divulgación y la formación científica y/o educativa.

<h2 align="center">Tipos de celdas</h2>


Esencialmente existen tres tipos de celdas:

- <a id="code">Celdas de código (Code)</a>
    > Por default la primera celda es del tipo "Code" y se inicia con el Kernel que se elija. En este tipo de celdas se puede ejecutar secuencias de códigos y/o programas completos generando. Estas celdas generalmente ejecutan una salida. Cualquier tipo de texto que no sea código debe ser precedido por el símbolo # para crear comentarios dentro de la celda y no pueden ser formateados como texto. 

- <a id="markdown">Celdas de marcado (Markdown)</a>
    > El segundo tipo de celda es "Markdown" y en ella podemos introducir texto y formatearlo, incluir imágenes, videos, enlaces, ecuaciones, entre otros elementos. Para formatear y diseñar el contenido de estas celdas se puede usar los lenguajes markdown o html.  

- <a id="raw">Celdas crudas (Raw)</a> 
    > En las celdas sin procesar cualquier contenido que contengan cuando se ejecuten permanecerá exactamente como está. Se utilizan principalmente para crear y conservar ejemplos de código o marcado de texto sin que éste se procese al ejecutarse la celda.

<h2 align="center">Consola</h2>
    
La *consola Jupyter* es una interfaz de terminal para núcleos que utilizan el protocolo Jupyter. Esta se ejecuta basada en el núcleo (kernel) en el que se esté ejecutando. En la consola podemos ejecutar comandos y código relacionado al [kernel](#Kernels) que estemos usando. En el caso de JupyterLab ésta se puede ejecutar desde la pestaña [Launcher](#inter_1) o desde un notebook dándo clic derecho en una celda y eligiendo "*New Console for Notebook*"; esto ejecutará una nueva pestaña con una consola donde podremos ejecutar comandos para operar sobre ese notebook.

<h2 align=center>Terminal</h2>
    
La *terminal Jupyter* proporciona un soporte completo para [shells](https://es.wikipedia.org/wiki/Shell_(inform%C3%A1tica)) del sistema (bash, tsch, etc.) en Mac / Linux y PowerShell en Windows. Puede ejecutar cualquier cosa en el shell de su sistema con un terminal, incluidos programas como vim o emacs. Los terminales se ejecutan en el sistema donde se ejecuta el servidor Jupyter, con los privilegios de su usuario. Por lo tanto, si JupyterLab está instalado en su máquina local, los terminales JupyterLab se ejecutarán allí. Se puede ejecutar desde la pestaña [Launcher](#inter_1)

# Lenguajes de marcado
<hr width=300px size=2, align="left">
<hr width=250px size=2, align="left">

Un lenguaje de marcado o lenguaje de marcas es una forma de codificar un documento que, junto con el texto, incorpora etiquetas o marcas que contienen información adicional acerca de la estructura del texto o su presentación.

Un ejemplo de cómo funciona el lenguaje de marcado puede observarse en el dictado de viva voz de un documento a una persona que lo transcribe a máquina:

*Ponga estilo de carta, ponga comillas, ponga mayúsculas, Estimado Juan, ponga dos puntos, aparte, sangría, ponga primera letra mayúscula, te escribo esta carta, ponga negrillas, de forma muy urgente, cierre negrilla, ya que no me has enviado...*, etc.

En otras palabras, nos permite "convertir" un texto plano en un texto formateado visualmente más presentable. Actualmente el lenguaje más extendido en el mundo es **HTML**, sin embargo, existen muchos otros como Markdown. 

Jupyter actualmente se basa en estos lenguajes de marcado para formatear texto, insertar elementos multimedia entre otras cosas. Veamos una breve descripción de cada uno de ellos. 

<h2 align=center>Markdown</h2>

Markdown es un lenguaje de marcado ligero que facilita la aplicación de formato a un texto empleando una serie de caracteres de una forma especial. Es similar al lenguaje de marcado de hipertexto (HTML) hecho de etiquetas de marcado, el cual consiste en etiquetas de apertura y cierre. Es un estándar de escritura para científicos y analistas de datos y debido a su implementación puede ser compartido fácilmente entre diferentes dispositivos y personas.

    <figcaption>Figura 6: Markdown</figcaption></center>

<h2 align=center>HTML</h2>

HTML es un lenguaje de marcado que se utiliza para el desarrollo de páginas de Internet. Se trata de las siglas que corresponden a *HyperText Markup Language (Lenguaje de Marcas de Hipertexto)*.

HTML es un lenguaje formado por elementos, un elemento se ve como todo el texto que se muestra en la Figura 2, los elementos nos ayudan a estructurar y dar significado a las partes de un documento HTML, estos elementos a su vez están conformados por contenido, etiquetas y atributos.

<center><a id='fig_7' href="ImDB/0_0.png"><img src="ImDB/0_0.png" width="80%" height="80%"></a>
              <figcaption>Figura 7: HTML</figcaption></center>
<br>              
              
Las etiquetas nos sirven para delimitar el inicio y el fin de un elemento, como en la imagen, vemos un elemento que utiliza la etiqueta de apertura para indicar el inicio de un párrafo, y la etiqueta de cierre para indicar el fin de un párrafo. En medio de las etiquetas se agrega el contenido al cual deseamos aplicar el procedimiento. 

Los atributos de un elemento *siempre* se expresan en la etiqueta de inicio (Ej. "id") junto con un valor (Ej. "parrafo").

# Escribiendo con Markdown & HTML
<hr width=460px size=2, align="left">
<hr width=410px size=2, align="left">


En esta sección veremos cómo agregar y formatear texto entre otros elementos en las celdas del tipo [markdown](#markdown). A diferencia de otros procesadores de texto como Google Docs, Microsoft Word o Write de Open Office en los cuales podemos formatear texto a través de "accesos directos", en el caso de las celdas Markdown de los notebook de Jupyter, para poder dar formato o estilo al texto plano que escribimos en ellas se usan los [lenguajes de marcado Markdown o HTML](#Lenguajes-de-Marcado). 

En las siguientes sub-secciones veremos distintos estilos de formato para texto, así como incluir imágenes, enlaces internos y externos, tablas, entre otras cosas útiles para documentar nuestro trabajo usando los dos lenguajes de marcado para comparar sus ventajas y desventajas en cada caso. 

<h2 align="center">Encabezados</h2>

Tanto en Markdown como en HTML se pueden crear encabezados (títulos), y en los cuales sólo existen *seis niveles de encabezados* donde el primer nivel corresponde al encabezado más grande y el sexto nivel (último) el encabezado más pequeño. 

En Markdown se usan almohadillas o numeral (`#`) para encabezados y el número de ellas determina el nivel del encabezado. Por otro lado, en HTML, los encabezados se denotan explícitamente a través de las etiquetas `<h1>,...,<h6>` donde la primera es el título más grande o primer nivel.

<div class="alert alert-warning">
    <strong>Advertencia!</strong>
    <b>Es importante hacer notar que la etiqueta de cierre en html <i>siempre</i> lleva una diagonal inversa antes del nombre del elemento. De no hacerlo, HTML lo interpreta como si el tag de apertura no esté cerrado.</b>

Adicionalmente podemos ver que una ventaja de usar Markdown frente a HTML para encabezados, es que el primero no necesita una almohadilla de cierre a diferencia de HTML donde sí es necesario abrir y cerrar el tag. 

Finalmente tenemos que señalar que **NO** es posible generar más de seis niveles de encabezados, es decir, no existe un séptimo nivel aunque coloquemos siete almohadillas o coloquemos un inexistente tag `<h7>`.

<h4 align="center">Subrayado y marcado</h4>

Para enfatizar de manera alternativa algún texto más allá de negritas o itálicas podemos hacer uso del lenguaje HTML para renderizar texto plano en los siguientes estilos:

```html
<u>Subrayado</u>
<ins>Subrayado</ins>
```
> <u>Subrayado</u>

```html
<mark>Marcado</mark>
```
> <mark>Marcado</mark>

<h3 align="center">Tamaño de letra</h3>

Referente al tamaño de letra existen tags especiales para ello en HTML. Los tags `<big>` y `<small>` incrementan o decrecen el tamaño de la fuente por un factor de uno: 

```html
<small>Pequeño</small>
```
> <small>Pequeño</small>

```html
<big>Grande</big>
```
> <big>Grande</big>

Una utilidad particular de los tags anteriores es que podemos usarlos repetidamente lo que permite aumentar o disminuir aún más el tamaño del texto. 

Por otro lado, podemos definir un tamaño de fuente más "controlado" a través del tag `<font>` y usando el parámetro `size=`

```html
<font size=1>Tamaño 1</font>
```
> <font size=1>Tamaño 1</font>

```html
<font size=7>Tamaño 7</font>
```
> <font size=7>Tamaño 7</font>

Por default en el caso de este tag el intervalo de posibles tamaños está entre 1 a 7. 

<h3 align="center">Color de texto</h3>

A través del tag `<font>` visto anteriormente también es posible cambiar el color del texto usando el parámetro `color=`, por ejemplo:

```html
<font color="red">Texto rojo</font>
```
> <font color="red">Texto rojo</font>

```html
<font color="blue">Texto azul</font>
```
> <font color="blue">Texto azul</font>

<h3 align="center">Alineación de texto</h3>

Si se necesita alinear un texto, ya sea "justificado", "a la izquierda", "a la derecha" o "centrado", se puede hacer a través del tag `<p>` para párrafos, usando el parámetro `style="text-align:`


```html
<p style="text-align:left">Texto alineado a la izquierda.</p>
```
> <p style="text-align:left">Texto alineado a la izquierda.</p>

```html
<p style="text-align:right">Texto alineado a la derecha.</p>
```
> <p style="text-align:right">Texto alineado a la derecha.</p>

```html
<p style="text-align:justify">Texto justificado.</p>
```
> <p style="text-align:justify">Texto justificado.</p>

```html
<p style="text-align:center">Texto centrado.</p>
```
> <p style="text-align:center">Texto centrado.</p>

<h3 align="center">Subíndices y superíndices</h3>

Los superíndices y subíndices son posibles a través de los siguientes tags:

```html
Soy un <sup>superíndice</sup>
```
> Soy un <sup>superíndice</sup>

```html
Soy un <sup>subíndice</sup>
```
> Soy un <sub>subíndice</sub>

<h3 align="center">Formato combinado</h3>

Finalmente, podemos emplear un [atributo global de HTML (`style=`)](https://www.w3schools.com/tags/ref_standardattributes.asp "Global attributes")  que en combinación con el tag `<span>` podemos formatear texto combinando varios estilos a la vez y en una sola línea. 

```html
<span 
style="color:green; 
text-decoration:underline;
font-style: italic;
font-weight: bold;
font-size: 20px;
background-color: #FFFF00"></span>
```

> <span style="color: green; text-decoration: underline; font-style: italic; font-weight: bold; font-size: 20px; background-color: #FFFF00;">Texto en negritas color verde, subrayado, con marcado amarillo y fuente 20px</span>

# Saltos de sección y línea
<hr width=320px size=2, align="left">
<hr width=270px size=2, align="left">

Podemos definir visualmente un salto de sección o de tema usando líneas horizontales. Tanto en Markdown como en HTML es posible crear una línea que vaya a lo largo de toda la celda markdown. En el caso de Markdown podemos crear una línea horizontal colocando tres o más "asteriscos", "guiones medios o bajos":

```markdown
---
___

***
```

---

También es posible usar HTML para crearlos a través del tag `<hr>` y personalizarlo usando parámetros como `align=`, `size` y `width`:

```html
<hr width=500px size=10, align=center>
```

<hr width=500px size=10, align=center>


Referente a los saltos de línea, en Markdown podemos realizarlo pulsando la tecla "Enter" para pasar a la siguiente línea y definir un nuevo párrafo o simplemente para dejar un espacio entre líneas. Empero, en HTML para poder hacerlo es necesario usar el tag `<br>` para definir un salto de línea en las página web.  

Dado que la celdas markdown en Jupyter primariamente trabajan con el lenguaje Markdown no es necesario entonces colocar el tag `<br>` para definir un salto de línea sino solamente tecleando un espacio, sin embargo, *existen ciertos casos* en los cuales al combinar HTML con Markdown puede tener resultados no deseados, y uno de ellos es que en ocasiones teclear no funciona para crear espacios en partes del documento. Es ahí, en esos "*errores*" donde se hace útil el tag `<br>` para definir uno o más saltos de línea. 

Cabe destacar que no es necesario incluir un tag de cierre para el tag `<br>` pues éste es único.

Tim Berners-Lee, *Tejiendo la Web*.

Crear tablas en Markdown resulta ser algo más sencillo en contraste con HTML. Pero aunque es más fácil crear tablas en Markdown, HTML le aventaja en opciones y diseño. Veamos la situación para cada caso.

<h2 align="center">Tablas en Markdown</h2>

Para crear tablas en Markdown se hace uso de la *pleca o barra vertical* `|`, y de guiones medios `-` para definir donde finzaliza el encabezado y dónde empieza los datos de la tabla. Veamos cómo se crea una tabla en Markdown con su respectiva salida renderizada:

<center>

```markdown
A|B|C
-|-|-
1|2|3
4|5|6
```
</center>
<br>
    
A|B|C
-|-|-
1|2|3
4|5|6
    
Como podemos ver para crear el encabezado introducimos los nombres de las columnas separados por una barra vertical. Luego, en la siguiente línea usamos el mismo número de guiones que de columnas para definir dónde termina el encabezado. Finalmente creamos las filas con la información que deseamos colocar separando cada celda de datos por una pleca y definiendo cada fila por un salto de línea. 
    
También podemos definir la alineación del encabezado y del contenido de las celdas. Por default está alineado "hacia la izquierda" pero podemos cambiarlo "hacia la derecha" o "centrado". Para definirlo usamos *dos puntos* `:` a los lados de los guiones que separan el encabezado de los datos para definir la alineación tanto del encabezado como de los datos.

<center>

```markdown
Izquierda|Centrado|Derecha
---------|:------:|------:
<<-|--|->>
Izq|Centro|Der
1|1|1
12|12|21
123|132|321
1234|1432|4321
```

Para crear la estructura general de tablas en HTML tenemos que hacer uso del tag `<table>`. Para la estructura del encabezado usamos `<thead>`, y para cada celda del encabezado usamos `<th>`. El cuerpo de la tabla de datos se define usando el tag `<tbody>`, las filas se definen usando el tag `<tr>`, y para cada celda de datos usamos `<td>`. # Enlaces
<hr width=100px size=2, align="left">
<hr width=50px size=2, align="left">


En las celdas de markdown es también posible crear "accesos directos", o mejor dicho, enlaces a otros sitios que pueden ser tanto "internos", es decir, en el mismo notebook, o "externos" para enlazar a algún sitio web o contenido específico en otro notebook. 

Hay una amplia variedad para crear enlaces usando el lenguaje de marcado Markdown. A continuación describimos distintas formas de crearlos.

<h2 align="center">Enlaces con Markdown</h2>


Para crear un enlace que conecte un "acceso directo" a algún recurso externo, por ejemplo, a una página web, se pueden generar de dos formas: colocando el enlace directamente o usando corchetes `[]` y paréntesis `()` para generar un de link descriptivo. 

- **Directo**: <br>

> https://www.algo.com/

- **Personalizado**: 

```markdown
[Enlace a algo](https://www.algo.com/)
```
> [Enlace a algo](https://www.algo.com/)

Si queremos crear un enlace que dirija a algún gestor de correo electrónico con la dirección de e-mail del destinatario podemos crearlo agregando el término `mailto:` dentro de los paréntesis seguido de la dirección de e-mail al cuál debe enlazar, por ejemplo:

```markdown
[Correo electrónico](mailto:user@mail.to)
```
> [Correo electrónico](mailto:user@mail.to)


- **Enlace personalizado con título**:

También es posible mostrar un "título" en el enlace al pasar el puntero del mouse sobre él, adhiriendo entre comillas el texto que queremos mostrar, por ejemplo:

```markdown
[Enlace a algo](https://www.algo.com/ "Mi enlace")
```
> [Enlace a algo](https://www.algo.com/ "Mi enlace")

- **Enlace referenciado con ID**:

Podemos referenciar un enlace sin necesidad de incluir el recurso directamente usando dos veces corchetes en lugar de paréntesis:

```markdown
Esto es un [enlace referido][id_referido]

[id_referido]: https://www.algo.com/ "Mi enlace"
```

> Esto es un [enlace referido][id_referido]

[id_referido]: https://www.algo.com/ "Mi enlace"

- **Enlace referenciado sin "ID"**:

Inlcuso es posible definir un enlace sin usar un "ID", simplemente usando el nombre de referencia se puede enlazar:

```markdown
Esto es un [enlace referido][]

[enlace referido]: https://www.algo.com/ "Enlace sin ID"
```
> Esto es un [enlace referido][]

[enlace referido]: https://www.algo.com/ "Enlace sin ID"

- **Enlace a secciones**

En el caso de los enlaces internos, en Markdown sólo se pueden ligar a secciones creadas mediante [encabezados](#Encabezados) usando la sintaxis anterior de enlaces. Por ejemplo, si queremos dirigirnos a una sección en específico, digamos, a la *Tabla de Contenidos* de nuestro notebook, tenemos que hacer uso de un numeral o almohadilla (`#`) e indicar el nombre de la sección tal cual aparece en el encabezado. 

Veamos el ejemplo a continuación:

```markdown
[Tabla de Contenidos](#Tabla-de-contenidos "Click aquí")
```

> [Tabla de Contenidos](#Tabla-de-contenidos "Click aquí")

<div class="alert alert-warning">
    <strong>Advertenia!</strong>
    <b>Tome en cuenta lo siguiente: si el nombre del encabezado tiene espacios entre palabras, es necesario sustituir esos espacios por guiones medios puesto que de no hacerlo, el enlace no funcionará.</b>
</div>


- **Enlace a otros notebooks**
    
Podemos crear un vínculo a otro notebook indicando el nombre del archivo junto con su extensión. Si no está abierto entonces Jupyter abrirá el archivo en una nueva instancia, y si ya está abierto simplemente nos dirigirá hacia él. 

Para enlazar a otro notebook, la sintaxis sería la siguiente:
    
```markdown
[Vínculo a otro notebook](nombre-de-notebook.ipynb)  
```
> [Vínculo a otro notebook](nombre-de-notebook.ipynb)

Nuevamente, si el nombre del archivo (notebook) contiene espacios entre palabras será necesario reemplazarlos por guines medios.
    
## Enlaces con HTML
    
En HTML para poder crear un link se usa el tag `<a>` (anchor) el cual nos permite vincular a un recurso externo o alguna parte específica un sitio web. Su funcionamiento en Jupyter no es muy distinto a Markdown pero con funciones adicionales. Al igual que en Markdown podemos crear un vínculo hacia un sitio web usando el atributo `href=` dentro del tag `<a>`. Adicionalmente podemos agregar un título al texto del vínculo para que al posar el puntero del mouse sobre el enlace muestre información.

- **Enlace personalizado**

```html
<a href="http://www.algo.com" title="Mi enlace">Link a web</a>
```
<a href="http://www.algo.com" title="Mi enlace">Link a web</a>

En HTML también podemos especificar una dirección de correo electrónico en lugar de una URL. Para realizarlo tenemos que especificar en `href=` el prefijo `mailto:` seguido de la dirección de correo electrónico que queremos especificar:

```html
<a href="mailto:user@mail.to">Correo electrónico</a>
```
<a href="mailto:user@mail.to">Correo electrónico</a>

- **Enlace a secciones**

Para crear enlaces internos a secciones, la estructura es similar a Markdown. Solo referenciamos a la sección que queremos usando su nombre y anteponiéndole una almohadilla (`#`), y sustituyendo los espacios por guiones medios.

```html
<a href="#Tabla-de-contenidos">Tabla de contenidos</a>
```
<a href="#Tabla-de-contenidos">Tabla de contenidos</a>

También podemos vincular a otros notebooks usando la misma estructura que en Markdown. Sólo tenemos que incluir en `href=` el nombre del notebook al cual nos queremos dirigir: 

```html
<a href="nombre-de-notebook.ipynb">Enlace a otro notebook</a>
```
<a href="nombre-de-notebook.ipynb">Enlace a otro notebook</a>
    
    
- **Enlace a otros elementos**

Una ventaja de crear enlaces con HTML frente a Markdown es que no sólo es posible vincular a secciones sino que podemos crear un *identificador único* usando otro atributo del tag `<a>` llamado `id=`. Este atributo se puede usar en casi cualquier etiqueta del código HTML incluso para referenciar imágenes o tablas. Veamos algunos ejemplos de cómo referenciar a distintos elementos usando este atributo:

> **Referenciando imágenes**

Cuando incluímos alguna [imágen](#Imágen) en nuestro notebook podemos recurrir al tag `<img>` para insertar la imagen pero además tenemos la posibilidad de especificar el atributo `id=` y asignar un identificador para posteriormente vincular a esa imagen desde otra parte del notebook usando un enlace. 

```html
 <img id='imagen_1' src="ImDB/0_0.png" width="10%" height="10%">
```

 <img id='imagen_1' src="ImDB/0_0.png" width="25%" height="25%">

En el ejemplo anterior, al parámetro `id=` le hemos asignado un *nombre único* para referenciar nuestra imágen y a través de ese identificador podemos crear enlaces con Markdown o HTML para vincular a esa imagen desde cualquier parte del notebook.

> **Referenciando tablas** 

Ya habíamos visto en la sección [Tablas](#Tablas) que podemos crearlas usando la etiqueta `<table>`. En este elemento también es posible usar el atributo `id=` para referenciar nuestras tablas:

```html
<table id="tabla_1"></table>
```
> **Referenciando párrafos** 

En Markdown para separar párrafos basta con introducir el texto y dar un salto de línea para pasar al siguiente párrafo.

En HTML, un párrafo se tienen que definir mediante el tag `<p>` necesariamente para diferenciar uno del otro. A pesar que en las celdas markdown no necesitamos especificar el tag `<p>` para separar párrafos, sí puede ser útil para referenciar un párrafo en específico a través del uso de `id=`. 

Como un ejemplo veamos el siguiente enlace a un párrafo en particular de este notebook:

```markdown
[Párrafo 2](#parr_2 "Introducción")
```
[Párrafo 2](#parr_2 "Introducción")

Como pudimos ver, existen distintas formar de crear vínculos o enlaces a distintas partes de nuestro notebook, a páginas web o incluso otros notebooks. Por supuesto, no son lás únicas formas. Existe una amplia variedad de enlazar a distintos objetos o ubicaciones mediante HTML. Por otro lado, Markdown a pesar de ser más accesible para la creación de enlaces resulta estar más limitado en funciones. 

<h2 align="center">¿Enlaces sin funcionar?</h2>

En algunas ocasiones los enlaces no funcionarán correctamente. Esto en parte sucede cuando el usuario escribe mal la ruta del enlace o no incluye la *almohadilla* en los enlaces internos. Pero habrá casos en los cuales escribamos correctamente la ruta del enlace, y no funcionará; esto es debido a que posiblemente el nombre del enlace contenta caracteres especiales que Makrdown o HTML no pueda interpretar de manera adecuada. 

Aunque Markdown regularmente no tiene problema en interpretar los caracteres especiales en los enlaces, en el caso particular de HTML sí representan un problema. 

Para comprobarlo veamos los dos siguientes con Markdown y HTML:

```markdown
[Enlace con caracteres especiales](#¿Enlaces-sin-funcionar?)
```
> Caso Markdown: [Enlace con caracteres especiales](#¿Enlaces-sin-funcionar?)

```html
<a href="#¿Enlaces-sin-funcionar?">Enlace con caracteres especiales</a>
```

> Caso HTML: <a href="#¿Enlaces-sin-funcionar?">Enlace con caracteres especiales</a>


Como podemos constatar de los dos casos anteriores, sólo el *caso Markdown* pudo vincular correctamente el enlace a la sección referida pero el *caso HTML* no tuvo el mismo éxito. Pero podemos solventar este problema codificando los caracteres especiales en una forma que internamente Jupyter lo entienda, y esto es posible a través de la [*codificación de URL o codificación pocentual*](https://www.w3schools.com/tags/ref_urlencode.asp "Enlace externo").

Consideremos el siguiente ejemplo: %41 representa el número 0x41 en [hexadecimal](https://es.wikipedia.org/wiki/Sistema_hexadecimal) (números en base 16) que en decimal (números en base 10) se traduce a 65 el cual es el código [ASCII](https://es.wikipedia.org/wiki/ASCII) de la letra A. Por lo tanto, la letra A podríamos codifircarla como %41 y en un enlace perfectamente podríamos enlazar el recurso ya sea usando Markdown o HTML. 

De esta manera entonces podemos codificar caracteres especiales. En nustro caso concreto, la sección tiene signos de interrogación que podemos codificar como: %C2%BF (¿) y %3F (?). 


<div class="alert alert-info">
    <strong>Información!</strong>
      Un detalle a condiderar con está codificación es que <b style="color:red">sustituir los espacios por su equivalente %20 NO funcionará</b> ya que aunque esté codificado tanto Markdown como HTML <b style="color:red">siguen interpretando ese caracter como un "espacio" y para que el enlace funcione los espacios tienen que ser sustituidos por guines medios</b>; esto quiere decir que <b>si codificamos el guion medio por su equivalente %2D entonces sí funcionará</b> porque al final sigue siendo un guion medio aunque esté codificado.
</div>

```markdown
[Enlace codificado](#%C2%BFEnlaces%2Dsin%2Dfuncionar%3F)
```
> [Enlace codificado](#%C2%BFEnlaces%2Dsin%2Dfuncionar%3F)

```html
<a href="#%C2%BFEnlaces%2Dsin%2Dfuncionar%3F">Enlace codificado</a>
```
> <a href="#%C2%BFEnlaces%2Dsin%2Dfuncionar%3F">Enlace codificado</a>

Finalmente nuestros enlaces fueron codificados de manera correcta y ahora funcionan los dos adecuadamente. Se recomienda que en la práctica se evite usar encabezados o nombre de "id" que contengan caracteres especiales para evitar complicaciones con los enlaces. 

En Internet hay varios recursos con tablas de conversión de caracteres especiales o incluso sitios que lo hacen de manera automática sólo introduciendo el caracter. Los siguientes enlaces pueden ser útiles para tales propósitos:

> [URL Enconder/Decoder](https://meyerweb.com/eric/tools/dencoder/)

> [HTML URL Encoding Reference](https://www.w3schools.com/tags/ref_urlencode.asp)

# Escape de caracteres
<hr width=280px size=2, align="left">
<hr width=230px size=2, align="left">


Un problema de trabajar con Markdown y caracteres especiales es que en ocasiones al querer hacer uso de algunos símbolos especiales como almohadillas (`#`) que son usados internamente por Markdown como funciones podemos obtener un resultado no deseado. Es decir, no siempre vamos desear que el uso de la almohadilla convierta una palabra en un *encabezado* o un guion medio convierta sea interpretado como un elemento de una lista. 

Para evitar que Markdown interprete los caracteres especiales como funciones, podemos hacer uso la *diagonal inversa/backslash* (`\`) para evitar la ejecución del renderizado. Veamos el siguiente ejemplo demostrando el uso de *backslash*:

\\#### Esto NO es un encabezado. 

#### Esto SI es un enzabezado 

Los caracteres de los cuales podemos omitir su comportamiento son los siguientes:

* \   Diagonal inversa
* `   Acento grave
* \*  Asterisco
* _   Guión bajo
* {}  Llaves
* []  Corchetes
* ()  Paréntesis
* \#  Numeral o almohadilla
* \+  Signo "más"
* \-  Signo "menos" / Guión medio 
* .   Punto
* !   Signo de exclamación
* |   Barra vertical o pleca

# Botones interactivos
<hr width=260px size=2, align="left">
<hr width=210px size=2, align="left">

Otro elemento de HTML que en la práctica no es muy usuado en notebooks de Jupyter son los "botones". Este elemento es clickeable, de manera que podemos incluir texto, imágenes o enlaces para acceder a ellos de forma interactiva. Por lo general se usan en combinación con enlaces para tener una forma más interactiva de acceder al recurso. 

Este elemento se define a través del tag `<button>` y cuenta con una seríe de atributos para distintos proprósitos pero para los objetivos de este manual esto queda fuera de su alcance.  

Como un ejemplo básico de su uso, lo vamos a emplear para crear un [enlace](#Enlaces) a la tabla de contenidos de nuestro notebook usando un botón en lugar de mostrar el texto que contiene el enlace:

```html
<button>[<span style="color:red; font-weight: bold; font-size: 10px">Tabla de Contenidos</span>](#Tabla-de-contenidos)</button>
```
<button>[<span style="color:red; font-weight: bold; font-size: 10px">Tabla de Contenidos</span>](#Tabla-de-contenidos)</button>

Como podemos ver aplicamos incluso algunos estilos adicionales al texto del botón combinando Markdown y HTML a la vez pero en la práctica se sugiere sólo usar HTML para fomatear texto con botones.

# LaTeX 
<hr width=80px size=2, align="left">
<hr width=50px size=2, align="left">


$\mathbf{L\!\!^{{}_{\scriptstyle A}} \!\!\!\!\!\;\; T\!_{\displaystyle E} \! X}$ (escrito LaTeX en texto plano) es un sistema de composición de textos, orientado a la creación de documentos escritos que presenten una alta calidad tipográfica. Por sus características y posibilidades, es usado de forma especialmente intensa en la generación de artículos y libros científicos que incluyen, entre otros elementos, expresiones matemáticas.

LaTeX está formado por un gran conjunto de macros de [TeX](https://es.wikipedia.org/wiki/TeX), escrito por [Leslie Lamport](https://es.wikipedia.org/wiki/Leslie_Lamport) en 1984, con la intención de facilitar el uso del lenguaje de composición tipográfica, ${\displaystyle \mathbf {T\!_{\displaystyle E}\!X}}$, creado por [Donald Knuth](https://es.wikipedia.org/wiki/Donald_Knuth). Es muy utilizado para la composición de artículos académicos, tesis y libros técnicos, dado que la calidad tipográfica de los documentos realizados en LaTeX, se considera adecuada a las necesidades de una editorial científica de primera línea, muchas de las cuales ya lo emplean.

Markdown también puede soportar la sintaxis de LaTeX para formatear ecuaciones matemáticas. Para renderizar las expresiones matemáticas Markdown usa la librería [MathJax](https://www.mathjax.org) de [JavaScript](https://www.javascript.com).

En este manual no ahondaremos en cómo crear documentos basados en LaTeX sino solamente en la sintaxis básica para mostrar notaciones matemáticas. 

Para mostrar ecuaciones o símbolos matemáticos en LaTeX, se usa el caracter (`$`). Por ejemplo, el siguiente código quedaría renderizado de la siguiente forma: 

```markdown
$y = x^2$
```
$y = x^2$

Podemos centrar la expresión matemática anterior encerrándola con doble `$`: 

```markdown
$$y = x^2$$
```
$$y = x^2$$

Por supuesto podemos crear expresiones más complejas como la siguiente:

```markdown
$$F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx$$
```
$$F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx$$

Incluso podemos aplicar comandos de formato para tener una forma más estructurada de las expresiones matemáticas, por ejemplo, como alinear varias ecuaciones: 

```markdown
\begin{align*}
P(A_1 \cup A_2 \cup A_3) ~ = ~ P(B \cup A_3) &= ~ P(B) + P(A_3) - P(BA_3) \\
&= ~ P(A_1) + P(A_2) - P(A_1A_2) + P(A_3) - P(A_1A_3 \cup A_2A_3)\\
&= ~ \sum_{i=1}^3 P(A_i) - \mathop{\sum \sum}_{1 \le i < j \le 3} P(A_iA_j) + P(A_1A_2A_3)
\end{align*}
```
\begin{align*}
P(A_1 \cup A_2 \cup A_3) ~ = ~ P(B \cup A_3) &= ~ P(B) + P(A_3) - P(BA_3) \\
&= ~ P(A_1) + P(A_2) - P(A_1A_2) + P(A_3) - P(A_1A_3 \cup A_2A_3)\\
&= ~ \sum_{i=1}^3 P(A_i) - \mathop{\sum \sum}_{1 \le i < j \le 3} P(A_iA_j) + P(A_1A_2A_3)
\end{align*}

# Notas finales del autor
<hr width=280px size=2, align="left">
<hr width=230px size=2, align="left">

Hasta aquí llega este manual de Markdown y HTML para notebooks de Jupyter. Como hemos visto a lo largo de las secciones, existen una serie de opciones y herramientas poderosas que nos permiten poder convertir un simple texto plano en un docomento estructurado de alta calidad, funcional y que podemos combinar con celdas de código donde podemos ejecutar bloques de código o incluso programas completos en diversos lenguajes. Además podemos compartir muy facilmente los notebooks creados a través de diversas plataformas y servicios en la nube. Incluso puede ser ejecutado en IDEs como [Visual Studio Code](https://code.visualstudio.com) y [Spyder](https://www.spyder-ide.org). 

Por supuesto Jupyter no es el único lugar donde se pueden crear notebooks. Existen otras opciones como Beaker Notebook, Apache Zeppelin, Spark Notebook, DataBricks Cloud, Rodeo IDE o nteract en los cuales también se pueden crear contenidos interactivos con texto enriquecido y código. Cada uno tienen sus ventajas y desventajas y el usuario es quien decide al final a cuál herramienta se adapta mejor. 

A pesar de lo laborioso que pueda llegar a ser el crear documentos en notebooks a través de los Lenguajes de Marcado, estos métodos ofrecen al usuario la ventaja de poder controlar la forma en cómo se presentan los elementos que conforman el documento y con calidad de publicación. Más aún, permite que el usuario aprenda a conceptualizar las *entradas* y *salidas* como dependientes una de otra. Esto le permite madurar conceptos relacionados a lenguajes de programación pues en resumidas cuentas, estas plataformas están diseñadas o dirigidas para aquellos usuarios que desean desarrollar sus habilidades en programación. 

Una de las ventajas de poder integrar varias tareas en una sola plataforma es que no necesitamos usar varios tipos de software al mismo tiempo para desarrollar un trabajo de programación y además documentarlo en otro. Otra ventaja es que es son lenguajes universales que pueden ser entendidos en diversas plataformas sin necesidad de realizar algún tipo de conversión adicional y que se apega fielmente a la idea que quiere transmitir el autor original. Es además transparente a los usuarios, es decir, cualquier persona puede fácilmente acceder al código interno que genera las salidas renderizadas en los notebooks y entender qué hizo el autor original pero además tienen la posibilidad de poder cambiarlo o adaptarlo a sus necesidades. 

Finalmente sólo queda agregar que la "*respuesta más simple a un problema puede ser su solución más elegante*". Con esto me refiero al hecho de que no siempre el documento más complejo necesariamente sea el más llamativo, entendible y accesible a todo el mundo. No es necesario siempre hacer uso de todos los detalles de Markdown o HTML para tener un documento legible. __Se puede hacer tan fácil o complicado como el usuario quiera pero hacer uso adecuado de estas herramientas hará que su trabajo sea accesible para todos.__

__Gracias.__
